Este informe presenta el análisis estadístico de las señales EEG del dataset EEG Motor Movement/Imagery procesadas para 109 voluntarios durante cinco condiciones experimentales: reposo , movimiento real de mano derecha , movimiento real de mano izquierda , imaginación de movimiento de mano derecha e imaginación de movimiento. Se extrajeron características espectrales en las bandas mu (8–13 Hz) y beta (13–30 Hz) para los canales C3, C4. 



## PRUEBAS DE HIPÓTESIS

Antes de aplicar las pruebas inferenciales, se verificó el supuesto de normalidad mediante el test de Shapiro-Wilk (con muestras ≤ 5000 observaciones). En todos los grupos evaluados el estadístico de Shapiro arrojó p < 0.001, rechazando la hipótesis de normalidad. Dado que:

1. Las distribuciones de potencia EEG son asimétricas positivas (sesgo a la derecha).
2. Las varianzas entre grupos son heterogéneas (especialmente en C4, donde la DE supera X 100  a la mediana).
3. El tamaño de los grupos es muy diferente (REST: n≈9 845 vs MI_R: n≈2 438).

Se optó por la prueba no paramétrica de Mann-Whitney U en lugar del t-test de Student. Este test compara las distribuciones de rango de dos grupos sin asumir normalidad ni igualdad de varianzas, siendo la alternativa apropiada al t-test cuando los supuestos paramétricos se violan.

El nivel de significancia adoptado es **α = 0.05** para todas las pruebas.


## RESULTADOS DE PRUEBAS ESTADÍSTICAS


Luego, con el fin de caracterizar la distribución de la potencia espectral en las bandas mu (8–12 Hz) y beta (13–30 Hz), se calcularon estadísticos descriptivos para cada condición experimental (reposo, movimiento real e imaginación de movimiento) en los canales de interés.

Se incluyeron medidas de tendencia central (media y mediana) y de dispersión (desviación estándar e IQR), dado que las señales EEG presentan alta variabilidad y puede haber presencia de valores atípicos. En este contexto, la media permite una visión general del comportamiento de los datos, mientras que la mediana y el rango intercuartílico (IQR) proporcionan una descripción más detallada frente a distribuciones no normales.

La inclusión de estas métricas es fundamental, ya que permite evaluar la posible asimetría de la distribución y detectar la presencia de outliers, lo cual influye directamente en la selección de pruebas estadísticas posteriores (paramétricas o no paramétricas).

## TIPO DE REPRESENTACIÓN GRÁFICA

Se eligió realizar diagramas de caja (boxplot) como representación principal por las siguientes razones:

**- Distribución no normal:** Las señales EEG presentan distribuciones sesgadas positivamente, con valores atípicos frecuentes producidos por artefactos residuales. El boxplot muestra la mediana y el IQR sin asumir normalidad.

**- Comparación simultánea:** Permite contrastar las cinco condiciones en un mismo panel, haciendo visualmente evidentes las diferencias en localización y dispersión.

**- Identificación de outliers:** Los puntos fuera de los bigotes revelan épocas con potencia anormalmente alta (posibles artefactos de movimiento), lo que informa sobre la calidad de los datos.

**- Complemento a la tabla:** La tabla ofrece valores exactos; el boxplot ofrece la forma de la distribución, información que la media y la DE no capturan en datos asimétricos.

Se generaron cuatro paneles (mu-C3, mu-C4, beta-C3, beta-C4) donde el eje X es la condición experimental y el eje Y es la potencia media en µV²/Hz.


## DIFERENCIAS ENTRE CONDICIONES

### Banda mu (8–13 Hz) — Canal C3



    
C3 registra la actividad de la corteza motora izquierda, en esta se controla contralateralmente la mano derecha. En reposo se observa la mayor potencia mediana (5.54 µV²/Hz), consistente con una sincronización cortical en ausencia de demanda motora. Al realizar o imaginar movimiento de la mano derecha (MM_R, MI_R), la potencia mu en C3 disminuye notablemente:

- REST → MM_R: reducción del **31 %** en la mediana (5.54 → 3.82 µV²/Hz)
- REST → MI_R: reducción del **22 %** (5.54 → 4.32 µV²/Hz)

Esta reducción corresponde al lo que se conoce como desincronización relacionada con eventos (ERD), que refleja la activación de la corteza motora contralateral. Es el mecanismo neurofisiológico esperado: cuando la corteza motora izquierda se activa para procesar el movimiento de la mano derecha, las neuronas dejan de oscilar sincrónicamente en mu, reduciendo la potencia de esa banda.

También se observa que el ERD es mayor en movimiento real (MM_R) que en imaginación (MI_R), lo que tiene sentido fisiológico: la ejecución física involucra circuitos adicionales de retroalimentación propioceptiva y salida motora eferente que amplifican la desincronización.


### Banda mu (8–13 Hz) — Canal C4



    
C4 controla la mano izquierda. El patrón es análogo al anterior pero desde el hemisferio derecho. La mediana en Reposo  (5.73 µV²/Hz) disminuye en MM_L (3.83 µV²/Hz, −33 %) y MI_L (4.29 µV²/Hz, −25 %). Esto confirma lateralización contralateral del ERD: la tarea de mano izquierda desincroniza C4 pero no C3.

Se observa que la desviación estándar y la media en C4 son muy elevadas (REST: media=28.45, DE=190.63) en comparación con la mediana (5.73), lo que indica la presencia de valores atípicos extremos que inflan la media aritmética. Esto refuerza la decisión de usar la mediana como medida de centralidad.


### Banda beta (13–30 Hz) — Canales C3 y C4




    
La banda beta sigue un patrón similar al mu pero con magnitudes más bajas. En C3:

- REST: mediana = 2.36 µV²/Hz
- MM_R: mediana = 1.67 µV²/Hz (−29 %)
- MI_R: mediana = 1.92 µV²/Hz (−19 %)

La disminución de beta durante el movimiento e imaginación es coherente con la literatura: la desincronización beta acompaña a la mu como marcador de activación cortical motora y es particularmente sensible al inicio del movimiento voluntario.

En la banda beta, se observa una reducción de la potencia durante las condiciones activas en comparación con el reposo, especialmente en el canal C3. Este comportamiento es consistente con la desincronización beta durante la ejecución del movimiento.

Aunque en este análisis no se segmentaron explícitamente ventanas post-movimiento, los resultados sugieren una posible dinámica compatible con el fenómeno de rebote beta (ERS), el cual se caracteriza por un aumento de potencia posterior a la actividad motora. Sin embargo, para confirmar este fenómeno sería necesario realizar un análisis temporal más detallado.

En el canal C4, nuevamente se evidencia una alta variabilidad en los datos, lo que puede estar asociado a ruido, artefactos o diferencias individuales entre sujetos.

### Comparación Movimiento Real vs Imaginación

    
En todas las bandas y canales, el movimiento real produce mayor ERD que la imaginación. La diferencia entre MM_R y MI_R en mu-C3 es del 12 % (3.82 vs 4.32 µV²/Hz). Esto sugiere que, aunque la imaginación motora activa circuitos similares, la activación es de menor magnitud, lo que constituye el fundamento neurofisiológico sobre el cual se apoyan los sistemas BCI: la imaginación genera señales diferenciables del reposo pero también distintas del movimiento real.

### Asimetría de la Actividad entre Hemisferios


Un hallazgo importante es que durante MI_R, la potencia mu en C3 (4.32 µV²/Hz) es menor que en C4 (4.98 µV²/Hz). Esta asimetría lateral es la manifestación del ERD contralateral y es precisamente la hipótesis H1 que se prueba formalmente en la sección siguiente.


## IMPLEMENTACIÓN DE MAPA DE CALOR




    
El uso de mapas de calor permite comparar de manera compacta la potencia espectral entre múltiples condiciones, canales y bandas de frecuencia. En este caso, se presentan dos enfoques:

Valores absolutos (µV²/Hz) - magnitud real de la potencia

Valores normalizados (z-score) - comparación relativa entre condiciones

Esta combinación es adecuada, porque los valores absolutos muestran la escala fisiológica real y el z-score permite identificar patrones relativos eliminando el sesgo de escala entre canales.

El mapa de calor permite visualizar las diferencias en la potencia espectral entre condiciones, canales y bandas de frecuencia. En los valores absolutos, se observó que la condición de reposo presenta la mayor potencia en la banda mu, lo cual es consistente con un estado de sincronización cortical. Por el contrario, las condiciones de movimiento real e imaginación mostraron una reducción de la potencia, evidenciando el fenómeno de desincronización relacionada con eventos (ERD).

El análisis mediante valores normalizados (z-score) permitió identificar patrones más claros entre condiciones, eliminando el sesgo de escala entre canales. Se evidenció que el movimiento real presenta los valores más negativos, indicando una mayor desincronización en comparación con la imaginación motora, la cual mostró valores intermedios. Este resultado es consistente con la literatura, donde se reporta que la ejecución real del movimiento genera una mayor activación cortical que su imaginación.

En la banda beta se observó un comportamiento similar, aunque menos pronunciado, lo cual es esperable dado que el rebote beta requiere un análisis temporal posterior al evento. En general, los resultados obtenidos respaldan parcialmente las hipótesis planteadas, especialmente en lo referente a la diferencia entre movimiento real e imaginación.




## RESULTADOS DE HIPOTESIS PLANTEADAS INICIALMENTE SOBRE LATERALIZACIÓN Y MODULACIÓN DE LA ACTIVIDAD CORTICAL MOTORA 

### Hipótesis 1: Lateralización mu durante Imaginación de Movimiento de Mano Derecha (MI_R)


**Fundamento neurofisiológico:** La imaginación del movimiento de la mano derecha debe activar preferentemente la corteza motora izquierda (registrada en C3). Si existe ERD contralateral, la potencia mu en C3 debería ser significativamente menor que en C4 durante MI_R.

**Planteamiento de hipótesis:**

Hipotesis Alternativa H1: Durante la imaginación de movimiento de la mano derecha (MI_R), la potencia del ritmo mu (8–13 Hz) en el canal C3 será significativamente menor que en el canal C4.

Hipotesis Nula H0: No existen diferencias significativas en la potencia del ritmo mu entre los canales C3 y C4 durante la condición MI_R.

**Prueba:** Mann-Whitney U unilateral (parte izquierda), ya que la hipótesis alternativa tiene dirección específica.


**Resultados obtenidos:**

| Estadístico | Valor |
|---|---|
| Mediana mu-C3 durante MI_R | 4.3209 µV²/Hz |
| Mediana mu-C4 durante MI_R | 4.9797 µV²/Hz |
| Estadístico U | 2 759 446.00 |
| p-valor | 0.000008 |
| Decisión (α = 0.05) | **Se rechaza H₀** |

    
**Interpretación:** Con p = 8×10⁻⁶ << 0.05, existe evidencia estadística para rechazar H₀. La potencia mu en C3 es significativamente menor que en C4 durante la imaginación de movimiento de mano derecha. Esto confirma la presencia del ERD contralateral en la banda mu, que muestra la activación de la corteza motora izquierda durante la imaginería motora de la mano derecha. Este resultado valida el fundamento fisiológico sobre el cual se construyen los sistemas BCI basados en imaginería motora.



### Hipótesis 2 : Diferencia entre reposo e imaginación en mano derecha - canal contralateral (REST vs MI_R)

**Fundamento neurofisiológico:** Si la imaginación de movimiento de mano derecha activa la corteza motora izquierda, debe producir una reducción de la potencia beta en C3 respecto al estado de reposo (ERD beta).

**Planteamiento de hipótesis:**

Hipotesis Alternativa H1: La potencia del ritmo beta (13-30 Hz) en el canal C3 será significativamente menor durante la imaginación de movimiento de la mano derecha (MI_R) en comparación con la condición de reposo.

Hipotesis Nula H0: No existen diferencias significativas en la potencia del ritmo beta en el canal C3 entre la condición de reposo y la imaginación de movimiento de la mano derecha.

**Prueba:** Mann-Whitney U bilateral (no se asume normalidad; distribuciones confirmadas como no normales por Shapiro-Wilk, p < 0.0001 en ambos grupos).

**Resultados obtenidos:**

| Estadístico | Valor |
|---|---|
| n REST | 9 845 épocas |
| n MI_R | 2 438 épocas |
| Mediana beta-C3 REST | 2.3646 µV²/Hz |
| Mediana beta-C3 MI_R | 1.9213 µV²/Hz |
| Shapiro p (REST / MI_R) | 0.0000 / 0.0000 → NO normal |
| Estadístico U | 13 361 879.00 |
| p-valor | ≈ 0.000000 |
| Decisión (α = 0.05) | **Se rechaza H₀** |

**Interpretación:** El p-valor es prácticamente nulo (p << 0.0001), lo que proporciona evidencia estadística abrumadora para rechazar la hipótesis nula. La potencia beta en C3 durante MI_R (mediana = 1.92 µV²/Hz) es significativamente menor que en REST (mediana = 2.36 µV²/Hz), con una reducción relativa del **19 %** en la mediana. Este resultado confirma que la imaginación de movimiento de mano derecha induce desincronización beta en la corteza motora contralateral (C3), un marcador neurológico bien establecido en la literatura de BCI. La diferencia, aunque estadísticamente robusta, tiene una magnitud moderada en términos absolutos, lo que es esperable dado que la imaginación produce activación cortical de menor intensidad que el movimiento real.


### Hipótesis 3 :Diferencia entre el movimiento real de la mano izquierda vs imaginación - canal contralateral : banda mu de C4 (MM_L vs MI_L)

<div style="text-align: justify;">

**Fundamento neurofisiológico:** Tanto el movimiento real como la imaginación de la mano izquierda activan la corteza motora derecha (C4). Sin embargo, se espera que el **movimiento real** produzca mayor desincronización mu (menor potencia) que la imaginación, dado que involucra retroalimentación sensoriomotora adicional.

**Planteamiento de hipótesis:**

Hipotesis Alternativa H1: La potencia del ritmo mu (8–13 Hz) en el canal C4 será significativamente menor durante el movimiento real de la mano izquierda (MM_L) en comparación con la imaginación de movimiento (MI_L), reflejando un ERD de mayor magnitud en la condición real.

Hipotesis Nula H0: No existen diferencias significativas en la potencia del ritmo mu en el canal C4 entre las condiciones de movimiento real e imaginación de la mano izquierda.

**Prueba:** Mann-Whitney U bilateral (Shapiro-Wilk p < 0.0001 en ambos grupos → prueba no paramétrica).

**Resultados obtenidos:**

| Estadístico | Valor |
|---|---|
| n MM_L | 2 471 épocas |
| n MI_L | 2 479 épocas |
| Mediana mu-C4 MM_L | 3.8318 µV²/Hz |
| Mediana mu-C4 MI_L | 4.2857 µV²/Hz |
| Shapiro p (MM_L / MI_L) | 0.0000 / 0.0000 → NO normal |
| Estadístico U | 2 859 451.00 |
| p-valor | 0.000052 |
| Decisión (α = 0.05) | **Se rechaza H₀** |

**Interpretación:** Con p = 5.2×10⁻⁵ << 0.05 se rechaza la hipótesis nula. La potencia mu en C4 es significativamente diferente entre movimiento real e imaginación de mano izquierda. La mediana es menor en MM_L (3.83) que en MI_L (4.29), confirmando que el movimiento real induce mayor ERD (mayor desincronización, menor potencia) que la imaginación. La diferencia porcentual es del ~11 %, lo que aunque modesta, es estadísticamente consistente con el hecho de que el movimiento físico involucra adicionalmente los circuitos espinales, la retroalimentación propioceptiva y la activación de neuronas espejo en la corteza somatosensorial, todos contribuyendo a una desincronización cortical más profunda.

## CONCLUSIONES

- En conjunto, los datos estadísticos descriptivos evidencian patrones consistentes con la literatura sobre actividad motora en EEG, particularmente la desincronización en las bandas mu y beta durante tareas de movimiento e imaginación.

- Las tres hipótesis planteadas fueron rechazadas con alta significancia estadística (p < 0.0001 en todos los casos), lo que proporciona evidencia sólida de que las diferencias observadas en la potencia espectral entre condiciones no son atribuibles al azar.

- La elección de la prueba no paramétrica Mann-Whitney U fue correcta y necesaria, dado que el test de Shapiro-Wilk rechazó la normalidad en todos los grupos evaluados. El uso de un t-test habría sido inapropiado y podría haber producido inferencias erróneas dada la marcada asimetría de las distribuciones.
  
- En la banda mu, se observa una disminución de la potencia en las condiciones activas (movimiento real e imaginación) en comparación con el estado de reposo, particularmente en el canal C3.Se observa como la reducción de la potencia en esta banda durante la activación motora.

- En el canal C3, asociado al control de la mano derecha, la condición de reposo presenta una media considerablemente mayor que las condiciones de movimiento e imaginación, lo que sugiere una activación cortical durante las tareas. Asimismo, las condiciones de movimiento real (MM) muestran valores ligeramente menores que las de imaginación (MI).
  
- En el canal C4 se da un comportamiento similar al C3, aunque con una dispersión significativamente mayor, que se ve en valores elevados de desviación estándar.
  
- La H1 confirmó el ERD contralateral en mu durante MI_R, la corteza motora izquierda (C3) se desincroniza más que la derecha (C4) al imaginar el movimiento de la mano derecha. 

- La H2 confirmó que la banda beta en C3 disminuye significativamente durante MI_R respecto a REST. Esto demuestra que la desincronización beta es un marcador accesible de la activación cortical motora incluso durante la imaginería, sin necesidad de ejecución física del movimiento.

- La H3 demostró que el movimiento real (MM_L) produce mayor ERD en C4 que la imaginación (MI_L), diferencia que es estadísticamente significativa aunque cuantitativamente moderada (~11 %). Esto tiene implicaciones prácticas importantes.
