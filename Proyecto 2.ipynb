**Common Spatial Pattern (CSP)**

El CSP es una técnica de filtrado espacial que mejora la capacidad discriminativa de las señales EEG al ser particularmente útil para problemas de clasificación binaria, como distinguir entre imaginación de movimiento de mano derecha e izquierda. El CSP opera encontrando filtros espaciales que maximizan la varianza para una clase mientras la minimizan para la otra. Este principio lo hace directamente complementario a los análisis PSD del Proyecto 1, donde la lateralización entre C3 y C4 fue el hallazgo central: mientras la PSD cuantifica la potencia en cada canal por separado, el CSP encuentra la combinación lineal de todos los canales que maximiza esa diferencia inter-hemisférica de forma óptima.[1]

Durante la imaginería motora, los ritmos sensoriomotores se atenúan y luego se amplifican en corto tiempo, fenómeno conocido como desincronización/sincronización relacionada con eventos (ERD/ERS). Para extraer de forma óptima las características EEG que describen este fenómeno, el algoritmo CSP busca filtros espaciales que extraigan características espaciales discriminativas entre clases y es adoptado frecuentemente debido a su buen desempeño. Sin embargo, el desempeño del CSP depende fuertemente de la selección de bandas de frecuencia para la extracción del ritmo sensoriomotor. Por ello, variantes como Filter Bank CSP (FBCSP) aplican el método sobre múltiples bandas de frecuencia antes de seleccionar las más discriminativas. La implementación en Python se realiza con mne.decoding.CSP de la librería MNE. [2]

**Parámetros de Hjorth**

Los parámetros de Hjorth son características basadas en la varianza de las derivadas de la señal EEG. Las tres primeras derivadas de la señal dan lugar a las medidas de actividad (Activity), movilidad (Mobility) y complejidad (Complexity), que son los parámetros de Hjorth frecuentemente empleados en sistemas BCI. Estos tres descriptores operan enteramente en el dominio del tiempo, lo que los hace ortogonales a la PSD y muy rápidos de calcular. [3]

Activity mide la varianza cuadrática de la amplitud de la señal EEG en una ventana considerada. Mobility se define como la potencia promedio de la derivada normalizada de la señal EEG. Complexity representa la relación entre las movilidades, es decir, la movilidad de la primera derivada dividida entre la movilidad de la señal original. En términos fisiológicos, Activity es análoga a la potencia total integrada de la PSD, Mobility refleja desplazamientos en la frecuencia media dominante del ritmo mu o beta, y Complexity captura cuánto se aleja la señal de una oscilación sinusoidal pura, lo que aumenta cuando la corteza motora se activa y la señal se vuelve más irregular. [4]

Según resultados recientes, la actividad, movilidad y complejidad de segmentos EEG captados por los canales C3, Cz y C4, con y sin descomposición en sub-bandas, proporcionaron precisiones de clasificación confiables. El enfoque propuesto permitió discriminar tareas de movimiento ejecutado e imaginado, alcanzando una precisión promedio de clasificación de 89.7 ± 0.78%. En estudios comparativos de extracción de características para BCI, se ha reportado que la combinación de potencia de bandas, parámetros de Hjorth y coeficientes autorregresivos adaptativos presenta resultados consistentes en múltiples datasets EEG de BCI. Los parámetros de Hjorth se implementan directamente con NumPy a partir de las derivadas discretas de la señal, sin necesidad de librerías especializadas. [5]

**Energía y Entropía Wavelet (DWT)**

La Transformada Wavelet Discreta (DWT) ofrece una representación tiempo-frecuencia de resolución múltiple que supera a la PSD convencional en señales no estacionarias como el EEG. Se propone un método de extracción de características basado en la transformada wavelet discreta (DWT), la descomposición en modos empíricos (EMD) y la entropía aproximada. La señal EEG se descompone en una serie de señales de banda estrecha con la DWT, y la entropía aproximada de la señal reconstruida se obtiene como el vector de características correspondiente. [5]

De la DWT se derivan dos índices especialmente útiles para BCI. El primero es la energía relativa por sub-banda, que representa la fracción de la energía total contenida en cada nivel wavelet, siendo directamente comparable con la PSD integrada por banda del Proyecto 1 pero con mejor localización temporal. El segundo es la entropía de Shannon wavelet, que mide qué tan distribuida está la energía entre las sub-bandas: la entropía wavelet relativa y las características topológicas de la red de función cerebral pueden extraerse de forma conjunta, logrando una precisión de clasificación promedio superior al 90% en datasets públicos de BCI, lo que demuestra que estas características retienen información de la señal EEG de forma más efectiva y reducen la complejidad computacional. [5]

En el proceso de extracción de características es posible calcular energía, varianza y entropía de la transformada wavelet basadas en cinco sub-bandas de frecuencia EEG como características en el dominio tiempo-frecuencia, las cuales, combinadas con parámetros en el dominio del tiempo y del dominio de la frecuencia, permiten obtener los mejores resultados de selección en clasificaciones binarias y multiclase de imaginería motora. La implementación en Python se realiza con la librería PyWavelets (pywt), que permite la descomposición DWT con wavelets como db4, cuyos niveles se alinean con las bandas delta, theta, mu y beta para una frecuencia de muestreo de 160 Hz.[6]

**Justificación de la selección**

Los tres índices presentados son complementarios entre sí y con respecto a la PSD: el CSP aporta información espacial que optimiza la detección del ERD contralateral entre hemisferios; los parámetros de Hjorth aportan información temporal sobre la velocidad y regularidad de la señal sin transformación al dominio frecuencial; y la entropía wavelet aporta información tiempo-frecuencial sobre la distribución y organización espectral de la energía. Para abordar los desafíos de la extracción de características y las bajas tasas de reconocimiento causadas por variaciones individuales en las señales EEG, los enfoques basados en fusión de múltiples características han demostrado mejorar la precisión de reconocimiento de señales EEG de imaginería motora, donde la combinación de características de energía, CSP y autorregresivas supera a cualquiera de ellas por separado.[7]

### **2.PLAN DE ANÁLISIS**

Con la información recolectada, del entregable 1 y del punto anterior, proponer una metodología de análisis que permita evidenciar la diferencia en ritmos cerebrales asociados a las condiciones de: reposo, imaginación de movimiento mano derecha, imaginación de movimiento mano izquierda. 

**1.Planteamiento del problema y justificación metodológica**

El objetivo central es discriminar tres condiciones **reposo (REST)**, **imaginación de movimiento de mano derecha (MI_R)** e **imaginación de movimiento de mano izquierda (MI_L)** , a partir de señales EEG del dataset PhysioNet EEG. El Proyecto 1 demostró que la PSD en las bandas mu (8–13 Hz) y beta (13–30 Hz) sobre los canales C3 y C4 produce diferencias estadísticamente significativas entre condiciones, confirmando el ERD contralateral como marcador neurofisiológico central. Sin embargo, la PSD por canal no tiene la información espacial entre hemisferios ni la estructura temporal de la señal. La siguiente metodología propuesta integra los tres índices adicionales identificados, CSP, parámetros de Hjorth y entropía wavelet para la clasificación por etapas, manteniendo la coherencia con los hallazgos del Proyecto 1.


**2. Selección de canales y condiciones**

Dado que el interés es la lateralización sensoriomotora entre las condiciones REST, MI_R y MI_L, la selección de canales no debe limitarse a C3 y C4 como en el Proyecto 1. Mediante la categorización de canales EEG por regiones cerebrales funcionales y la fusión de múltiples tipos de características, es posible mejorar la precisión, robustez e interpretabilidad de los sistemas BCI, donde los canales se agrupan según su relevancia funcional para regiones cerebrales específicas. Para este problema se propone trabajar con la región sensoriomotora ampliada: FC3, FC4 (corteza premotora), C3, Cz, C4 (corteza motora primaria) y CP3, CP4 (corteza somatosensorial), que son los canales donde el ERD/ERS mu y beta es más pronunciado durante la imaginería motora. 

Las tres condiciones se extraen de los runs correspondientes en el dataset PhysioNet: los runs 4, 8 y 12 contienen las tareas de imaginación (T1 = MI_R, T2 = MI_L), y el evento T0 corresponde a reposo. Se construyen épocas de 2 segundos con un margen de 0.5 s post-onset para evitar el potencial evocado de inicio de tarea, resultando en ventanas de análisis [0.5, 2.5] s, igual que en el Proyecto 1.

**3. Preprocesamiento**

El preprocesamiento sigue el mismo esquema del Proyecto 1 para garantizar comparabilidad: filtrado paso-banda 8–30 Hz (bandas mu y beta) con filtro FIR de fase cero, referencia al promedio común (CAR) para reducir artefactos de volumen conductor, y exclusión de épocas con amplitud superior a ±100 µV. El uso de filtrado de referencia promedio común (CAR) y filtrado Laplaciano de las señales EEG mejora la representación local de la actividad cortical antes de la extracción de características, constituyendo un paso de preprocesamiento innovador para la clasificación MI. Este paso es especialmente relevante para el CSP, que es sensible a la referencia espacial utilizada. 

**4. Extracción de características multidominio**

Esta es la etapa central de la metodología propuesta. Se calculan los tres índices sobre cada época y canal, construyendo un vector de características por época que combina información espectral (heredada del Proyecto 1), espacial, temporal y tiempo-frecuencial.

**4.1 PSD (línea base del Proyecto 1)**. Se conserva la potencia mediana en las bandas mu y beta para C3 y C4, calculada con el método de Welch, como referencia comparativa. Esto permite evaluar cuánto mejora la clasificación al añadir los nuevos índices.

**4.2 CSP (dominio espacial)**. El algoritmo CSP busca filtros espaciales para extraer características espaciales discriminativas entre clases. Sin embargo, el desempeño del CSP se ve fuertemente afectado por las bandas de frecuencia seleccionadas para la extracción del ritmo sensoriomotor y el período de tiempo utilizado para el análisis en la señal EEG. Por esta razón se propone entrenar un CSP por par de clases (REST vs MI_R, REST vs MI_L, MI_R vs MI_L) usando los 7 canales propuestos, extrayendo las log-varianzas de los 6 filtros más discriminativos (3 por clase) como características. Este esquema uno-contra-uno es coherente con las tres hipótesis del Proyecto 1, que también compararon pares de condiciones. 

**4.3 Parámetros de Hjorth (dominio temporal)**. Los parámetros de Hjorth para las señales EEG en reposo, incluyendo actividad, movilidad y complejidad, revelan agrupamientos de comportamiento distintos correlacionados con tareas motoras o estados mentales específicos, donde cada parámetro aporta información sobre la varianza de la señal, su frecuencia y la similitud de su forma de onda con una sinusoide pura. Se calculan los tres parámetros para C3 y C4, sobre la señal filtrada en banda mu y en banda beta por separado, generando 12 características (3 parámetros × 2 canales × 2 bandas). La Activity es directamente comparable con la PSD del Proyecto 1; Mobility y Complexity son información nueva.

**4.4 Entropía Wavelet (dominio tiempo-frecuencial)**. Las características en el dominio tiempo-frecuencial incluyen energía, varianza y entropía de la transformada wavelet basadas en cinco sub-bandas de frecuencia EEG, las cuales en procesos de selección estadística de características han demostrado proporcionar las mayores tasas de éxito en clasificaciones tanto binarias como multiclase de imaginería motora. Se aplica DWT con wavelet db4 a 5 niveles sobre C3 y C4; los niveles 3 y 4 corresponden aproximadamente a las bandas beta y mu para fs = 160 Hz. Se extraen la energía relativa por nivel y la entropía de Shannon wavelet por canal, generando 12 características adicionales (5 energías + 1 entropía × 2 canales). 


**5. Análisis estadístico diferencial**

Para determinar las características estadísticamente significativas, el uso de pruebas de hipótesis como el t-test independiente y el análisis de varianza ANOVA sobre las características extraídas permite identificar la mejor combinación posible entre todas las opciones de características para clasificaciones binarias y multiclase de imaginería motora. Dado que el Proyecto 1 estableció que las distribuciones EEG son no normales, se aplica la prueba de Kruskal-Wallis sobre cada característica, seguida de comparaciones post-hoc con Mann-Whitney U con corrección de Bonferroni para los tres pares (REST vs MI_R, REST vs MI_L, MI_R vs MI_L).
