# Descripción del Proyecto:

El dataset utilizado corresponde al EEG Motor Movement/Imagery Dataset, publicado en OpenNeuro. Este conjunto de datos contiene registros de EEG de 64 canales provenientes de 109 voluntarios que realizaron tareas motoras e imaginación de movimiento, adquiridos con el sistema BCI2000. Cada sujeto completó 14 corridas experimentales: dos líneas de base de un minuto (una con ojos abiertos, otra con ojos cerrados), y tres corridas de dos minutos para cada una de las cuatro tareas.[1]

Las cuatro condiciones experimentales relevantes para el curso son:

Movimiento real de la mano derecha (apertura/cierre del puño al aparecer el target)
Movimiento real de la mano izquierda
Imaginación de movimiento de la mano derecha (sin mover físicamente)
Imaginación de movimiento de la mano izquierda

Además existe una condición de reposo, que sirve como línea de base para los análisis de potencia espectral y coherencia.
Los EEG fueron registrados con 64 electrodos según el sistema internacional 10-10 (excluyendo Nz, F9, F10, FT9, FT10, A1, A2, TP9, TP10, P9 y P10). La cita principal del dataset es: Schalk, G. et al. BCI2000: A General-Purpose Brain-Computer Interface (BCI) System. 

El electroencefalograma (EEG) es una técnica de registro de la actividad eléctrica cerebral con alta resolución temporal, ampliamente utilizada en el estudio de interfaces cerebro-computadora (BCI, por sus siglas en inglés). Durante la imaginería motora, definida como la representación mental de un movimiento sin ejecución física del mismo, se producen cambios característicos en los ritmos cerebrales que pueden ser captados mediante EEG.16 motora.


# Historia

La historia de las BCI se remonta al siglo XIX con las primeras investigaciones en neurofisiología y el desarrollo del EEG. Con el tiempo, estas interfaces evolucionaron desde estudios puramente teóricos hasta aplicaciones reales que convierten impulsos cerebrales en comandos ejecutables. ScienceDirect
Los hitos más importantes en esta historia son:

1929: Hans Berger describe por primera vez el EEG humano y los ritmos alfa.
Años 70-90: Trabajos pioneros de Wolpaw, Pfurtscheller y Birbaumer demuestran el control de cursores en computadora mediante ritmos sensoriomotores.
2004: Publicación del sistema BCI2000 (Schalk et al.), que es precisamente la plataforma usada para adquirir el dataset de este proyecto.
2021: Un equipo de Stanford reportó que un participante cuadriplégico pudo producir oraciones en inglés a aproximadamente 18 palabras por minuto imaginando escribir letras a mano, mientras el sistema reconocía las señales eléctricas del córtex motor. Wikipedia

**Uso Actual**
Las BCI sintetizan ingeniería biomédica, neurociencia, ciencias de la computación e inteligencia artificial para analizar datos neurales y desarrollar algoritmos, mientras que la ciencia de materiales contribuye a mejorar la seguridad y estabilidad de los implantes. PubMed Central
Las aplicaciones actuales más importantes incluyen:

Rehabilitación motora post-ictus: En 2021 se reportó que sistemas BCI basados en EEG tienen eficacia para mejorar la función motora del miembro superior en comparación con terapias de control. Wikipedia
Comunicación aumentativa para pacientes con ELA o locked-in syndrome.
Control de prótesis, exoesqueletos y sillas de ruedas.
Entretenimiento y videojuegos neuro-controlados.

**Proyectos Futuros**

Las trayectorias futuras de investigación en BCI apuntan a desarrollos en tecnologías inalámbricas y totalmente implantables, buscando mecanismos de control más flexibles. Condiciones como ictus, lesión medular, depresión, TDAH, autismo y ELA pueden beneficiarse cada vez más de BCIs invasivas o no invasivas. Wiley Online Library
En el futuro se espera que las BCI se vuelvan más naturales, intuitivas y adaptables, permitiendo a los usuarios controlar dispositivos con entrenamiento mínimo. También se proyecta la integración con realidad aumentada y entornos de realidad virtual controlados por el pensamiento. ELEKS
En términos de mercado, el mercado global de BCI fue valorado en 2.1 billones de dólares en 2023, con proyecciones de alcanzar 4.5 billones para 2029, creciendo a una tasa anual del 14.2%. Bccresearch


# Actividad cerebral registrada por EEG
**Actividad Cerebral en EEG Durante Movimiento e Imaginación de Movimiento**

*Base Neurofisiológica* : Cuando una persona ejecuta o imagina un movimiento, ocurren cambios específicos y medibles en la actividad oscilatoria del EEG. Los dos fenómenos clave son:

- ERD (Event-Related Desynchronization): Reducción de potencia en una banda de frecuencia. Indica activación cortical (neuronas desincronizadas, procesando información activamente).
- ERS (Event-Related Synchronization): Aumento de potencia. Indica inhibición o "idle state" cortical.
  
Cuando los neurones desincronizadores reducen la amplitud de la oscilación EEG resultante, este fenómeno se llama ERD. Cuando la actividad de las neuronas se sincroniza, la amplitud aumenta, fenómeno llamado ERS. Estas representaciones en el dominio temporal pueden observarse como cambios de potencia en el dominio espectral. ScienceDirect
Ritmos Cerebrales de Interés
Las bandas de frecuencia más importantes en el contexto de BCI basadas en imaginación de movimiento son: mu (9-13 Hz), alfa (8-12 Hz), beta (13-30 Hz), y gamma (>30 Hz). Los ritmos mu, también conocidos como ritmos sensoriomotores, se encuentran en las partes del cerebro que controlan el movimiento voluntario. ScienceDirect
Los comportamientos específicos de cada ritmo son:
Ritmo Mu (8–13 Hz):

El ritmo mu relativo a las entradas sensoriomotoras está topográficamente restringido a las regiones C3 y C4, y por su especificidad fue denominado frecuencia Mu. ScienceDirect
Pfurtscheller et al. registraron oscilaciones mu durante cuatro tareas de imaginación cinestésica: mano derecha, mano izquierda, ambos pies y lengua. Mostraron ERD de mu en C3 y C4 durante ambas condiciones de imaginación de mano, con dominancia contralateral evidente. ScienceDirect

Ritmo Beta (13–30 Hz):

Se sabe que la imaginación de mano izquierda o derecha por un sujeto sano resulta en una desincronización (ERD) del ritmo mu (8-13 Hz) en el EEG contralateral, junto con sincronización (ERS) del ritmo beta (18-24 Hz) en el EEG ipsilateral. Springer

*Electrodos Relevantes*: Los electrodos más importantes para capturar movimiento/imaginación de manos son C3, Cz y C4, ubicados sobre el córtex sensoriomotor. La lógica es contralateral:
Para la imaginación de la mano izquierda, se observa ERD sostenido contralateral de mu y beta en C4 durante la imaginación, seguido de un rebote beta. La imaginación de la mano derecha muestra un patrón especular. ScienceDirect
Cuando se imagina el agarre con la mano izquierda, se observa un incremento de amplitud en el electrodo C3 y una disminución en C4, mientras que imaginar el agarre con la mano derecha muestra el patrón opuesto. Nature
Adicionalmente se incluyen FC3, FC4, CP3, CP4 como electrodos secundarios de la región pericentral.

# Flujos de Procesamiento Propuestos para Sistemas BCI

El estándar para análisis de imaginación de movimiento con EEG consta de las siguientes etapas:

Señal Raw EEG (64 canales)
        ↓
1. PREPROCESAMIENTO
   - Filtrado pasa-banda (0.5–40 Hz)
   - Filtro notch (60 Hz, para eliminar ruido de red)
   - Rereferenciación (promedio común o mastoides)
   - Remoción de artefactos: ICA (artefactos oculares/musculares)
        ↓
2. SEGMENTACIÓN (Epoching)
   - Alineación a markers de eventos
   - Extracción de épocas: ej. [-1, 4] segundos respecto al cue
   - Definición de línea de base (ventana pre-estímulo)
        ↓
3. EXTRACCIÓN DE CARACTERÍSTICAS
   - Análisis tiempo-frecuencia (STFT, wavelets de Morlet)
   - Potencia de bandas: mu (8-13 Hz), beta (13-30 Hz)
   - Cálculo de ERD/ERS por canal
   - CSP (Common Spatial Patterns) para maximizar discriminabilidad
        ↓
4. CLASIFICACIÓN
   - LDA, SVM, KNN, redes neuronales
   - Validación cruzada (k-fold)
   - Evaluación: accuracy, kappa de Cohen

La coherencia EEG provee una estimación importante de las interacciones funcionales entre sistemas neurales en cada banda de frecuencia y es frecuentemente utilizada para evaluar la conectividad funcional del córtex humano. Frontiers
La conectividad cerebral, particularmente la funcional y la efectiva, ha sido descrita como uno de los enfoques más prometedores para mejorar el rendimiento de los sistemas MI-BCI, ya que permite ir más allá de las características de un solo canal. ACM Digital Library
Tipos de conectividad utilizados en BCI:
Conectividad Funcional (no direccional):

Coherencia espectral por bandas (mu, beta)
Phase Locking Value (PLV)
Correlación cruzada

Conectividad Efectiva (direccional):

Causalidad de Granger, que incluye métodos como la Coherencia Parcial Dirigida Generalizada (GPDC), la Función de Transferencia Dirigida (DTF) y la dDTF, determinando la relación causal transitoria entre diferentes áreas cerebrales. PubMed Central

Tanto la conectividad efectiva como la funcional han sido investigadas en MI-BCI, empleando diversas métricas que incluyen coherencia, sincronización de fase, e índice de pendiente de fase, los cuales emplean diferentes algoritmos y difieren en su interpretación. PubMed Central
