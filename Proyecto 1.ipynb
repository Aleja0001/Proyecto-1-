#### *1.3 Actividad cerebral registrada por EEG - Actividad Cerebral en EEG Durante Movimiento e Imaginación de Movimiento*
<div style="text-align: justify;"> 

**Base Neurofisiológica** : Cuando una persona ejecuta o imagina ejecutar un movimiento, ocurren cambios específicos y medibles en la actividad oscilatoria del EEG. La ejecución y la imaginación de movimientos generan patrones específicos de actividad en la corteza motora del cerebro, estudios en neurociencia han demostrado que ambas condiciones activan regiones similares, lo que permite utilizar la imaginación motora como una alternativa viable en sistemas BCI.

En particular, durante estas tareas se observan cambios en los ritmos cerebrales registrados mediante EEG, especialmente en las bandas de frecuencia conocidas como ritmo mu (8–13 Hz) y ritmo beta (13–30 Hz). Estos ritmos están asociados con la planificación y ejecución del movimiento.[6]

Un fenómeno característico en este contexto es la desincronización relacionada con eventos (ERD), que se manifiesta como una disminución en la potencia de estas bandas durante la ejecución o imaginación del movimiento. Posteriormente, puede observarse una sincronización (ERS), asociada al retorno al estado de reposo.

- ERD (Event-Related Desynchronization): Reducción de potencia en una banda de frecuencia. Indica activación cortical (neuronas desincronizadas, procesando información activamente).
- ERS (Event-Related Synchronization): Aumento de potencia. Indica inhibición o "idle state" cortical.

Se sabe que la imaginación de mano izquierda o derecha por un sujeto sano resulta en una desincronización (ERD) del ritmo mu (8-13 Hz) en el EEG contralateral, junto con sincronización (ERS) del ritmo beta (13-30 Hz) en el EEG ipsilateral. 

**Electrodos Relevantes:**

El registro de la actividad EEG se realiza mediante electrodos colocados sobre el cuero cabelludo siguiendo el sistema internacional 10-10, el cual define ubicaciones estandarizadas para la medición de señales cerebrales.

En el caso del estudio de movimientos de las manos, los electrodos más relevantes son aquellos ubicados sobre la corteza motora:

C3: Asociado principalmente al control de la mano derecha

C4: Asociado al control de la mano izquierda

Cz: Ubicado en la región central

Estos electrodos permiten capturar la actividad neuronal relacionada con la planificación y ejecución del movimiento, siendo esenciales para el análisis en sistemas BCI.

#### *1.5 Coherencia y Conectividad Cerebral en BCI*

<div style="text-align: justify;"> 
En estudios más avanzados de BCI, se analiza la conectividad funcional entre diferentes regiones del cerebro. Una de las métricas más utilizadas es la coherencia, la cual mide el grado de sincronización entre señales EEG registradas en distintos electrodos.

El análisis de coherencia permite identificar redes neuronales que se activan de manera coordinada durante tareas específicas, como el movimiento o la imaginación motora. Esto proporciona información adicional que puede mejorar el rendimiento de los sistemas BCI. [7]

La coherencia espectral es una medida de conectividad funcional entre pares de canales EEG que indica qué tan sincronizados oscilan dos regiones cerebrales en una banda de frecuencia dada. Se define como:

$$C_{xy}(f) = \frac{|S_{xy}(f)|^2}{S_{xx}(f) \cdot S_{yy}(f)}$$

Donde $Sxy$​ es la densidad espectral cruzada y $S_{xx}$, $S_{yy}$ son las densidades espectrales de potencia de cada señal. Su valor oscila entre 0 (ninguna relación) y 1 (sincronización perfecta).

La coherencia EEG provee una estimación importante de las interacciones funcionales entre sistemas neurales en cada banda de frecuencia y es frecuentemente utilizada para evaluar la conectividad funcional del córtex humano. 

La conectividad cerebral, particularmente la funcional y la efectiva, ha sido descrita como uno de los enfoques más prometedores para mejorar el rendimiento de los sistemas MI-BCI, ya que permite ir más allá de las características de un solo canal. [7]

**Tipos de conectividad utilizados en BCI:**

- Conectividad Funcional (no direccional):
    Coherencia espectral por bandas (mu, beta)
    Phase Locking Value (PLV)
    Correlación cruzada

- Conectividad Efectiva (direccional):
    Causalidad de Granger, que incluye métodos como la Coherencia Parcial Dirigida Generalizada (GPDC), la Función de Transferencia Dirigida (DTF) y la dDTF, determinando la relación causal transitoria entre diferentes áreas cerebrales.

Tanto la conectividad efectiva como la funcional han sido investigadas en MI-BCI, empleando diversas métricas que incluyen coherencia, sincronización de fase, e índice de pendiente de fase, los cuales emplean diferentes algoritmos y difieren en su interpretación.[7]

   Intervalo: [-1 s, 4 s] respecto al estímulo

**3. EXTRACCIÓN DE CARACTERÍSTICAS:** En esta etapa se obtienen representaciones cuantitativas de la señal EEG que permiten comparar condiciones.

**-> Definición de la función para cargar los datos:**

````python
# ============================================================
# FUNCIÓN 1 — CARGA DEL ARCHIVO EEG CON EVENTOS
# ============================================================
# Recibe la ruta de un archivo .edf y retorna el objeto archivo
# de MNE con los datos cargados en memoria, los nombres de
# canales estandarizados y el montaje 10-20 asignado.
#
# ¿Por qué una función separada para cargar?
#   - Centraliza toda la lógica de carga en un solo lugar
#   - Si cambia el dataset o el formato, solo se modifica aquí
#   - Permite usarla dentro del pipeline masivo (todos los sujetos)
# ============================================================

def cargar_eeg_con_eventos(ruta_edf):
    """
    Carga un archivo EDF del dataset PhysioNet EEGBCI.

    Las anotaciones T0/T1/T2 vienen embebidas en el EDF+:
        T0 → reposo
        T1 → imaginación / movimiento mano izquierda
        T2 → imaginación / movimiento mano derecha

    Parámetros
    ----------
    ruta_edf : str
        Ruta absoluta o relativa al archivo .edf

    Retorna
    -------
    archivo : mne.io.Raw
        Objeto con la señal cargada, canales estandarizados
        y montaje 10-20 asignado.
    """

    # --- Carga del archivo EDF ---
    # preload=True: carga todos los datos en RAM.
    # Es necesario para poder filtrar y manipular la señal después.
    # verbose=False: suprime los mensajes internos de MNE.
    archivo = mne.io.read_raw_edf(ruta_edf, preload=True, verbose=False)

    # --- Estandarización de nombres de canales ---
    # El dataset PhysioNet usa nombres con puntos y mayúsculas
    # inconsistentes, por ejemplo 'Fc5.' en lugar de 'FC5'.
    # Esta función los convierte al estándar 10-20 reconocido por MNE.
    mne.datasets.eegbci.standardize(archivo)

    # --- Asignación del montaje estándar 10-20 ---
    # El montaje le indica a MNE la posición 3D de cada electrodo
    # sobre la cabeza. Es necesario para:
    #   - Graficar topomapas
    #   - Calcular la referencia promedio correctamente
    #   - Visualizar la PSD por canal
    montage = mne.channels.make_standard_montage('standard_1020')
    archivo.set_montage(montage, verbose=False)

    # --- Resumen en consola ---
    nombre = os.path.basename(ruta_edf)
    print(f"✓ Cargado : {nombre}")
    print(f"  Canales : {archivo.info['nchan']}  |  "
          f"sfreq : {archivo.info['sfreq']} Hz  |  "
          f"Duración : {archivo.times[-1]:.1f} s")
    print(f"  Anotaciones : {len(archivo.annotations)}  "
          f"→ {set(archivo.annotations.description)}")

    return archivo
````

**3.1 Función que recibe una señal de EEG y le aplica el flujo de procesamiento definido.**

````python
# ============================================================
# FUNCIÓN 2 — PREPROCESAMIENTO EEG POR ETAPAS
# ============================================================
# Aplica el flujo completo de preprocesamiento a una señal EEG
# y genera 3 figuras que muestran cómo cambia la señal en
# cada etapa. 
#
# ETAPAS APLICADAS:
#   0. Señal original (sin filtrar)         — línea de base
#   1. Filtro pasa-banda (1–40 Hz)          — elimina deriva y EMG
#   2. Filtro notch (60 Hz)                 — elimina red eléctrica
#   3. Rereferenciación CAR                 — reduce ruido común
#
# FIGURAS GENERADAS:
#   Fig 1 — Las 4 etapas superpuestas por canal (C3, C4, Cz)
#   Fig 2 — Original vs Procesada por canal
#   Fig 3 — PSD antes vs después (verifica eliminación de 60 Hz)
# ============================================================

def procesar_eeg(archivo, graficar=True, t_max_plot=30.0):
    """
    Aplica el flujo de preprocesamiento a una señal EEG cruda.

    Parámetros
    ----------
    archivo : mne.io.archivo
        Señal EEG cruda cargada con cargar_eeg_con_eventos().
    graficar : bool
        Si True, genera las 3 figuras comparativas.
        Si False, solo procesa y retorna la señal (útil para
        el pipeline masivo donde no se quiere visualizar cada archivo).
    t_max_plot : float
        Segundos a mostrar en las figuras (por defecto 30 s).

    Retorna
    -------
    archivo_proc : mne.io.archivo
        Señal EEG con los 3 filtros aplicados, lista para
        la extracción de características.
    """

    # ----------------------------------------------------------
    # CANALES DE INTERÉS
    # ----------------------------------------------------------
    # C3 → corteza motora izquierda (controla mano derecha)
    # C4 → corteza motora derecha   (controla mano izquierda)
    # Cz → región central, útil como referencia comparativa
    # Se filtra con 'if ch in archivo.ch_names' por si algún archivo
    # no tuviera alguno de estos canales (robustez).
    # ----------------------------------------------------------
    canales = [ch for ch in ['C3', 'C4', 'Cz'] if ch in archivo.ch_names]

    if not canales:
        raise ValueError(
            "El archivo EEG no contiene ninguno de los canales "
            "requeridos (C3, C4, Cz)."
        )

    # ----------------------------------------------------------
    # GUARDAR SEÑAL ORIGINAL ANTES DE MODIFICARLA
    # ----------------------------------------------------------
    # archivo.get_data() devuelve un array de forma (n_canales, n_muestras)
    # en voltios. Se guarda una copia antes de aplicar cualquier filtro
    # para poder comparar visualmente original vs procesada.
    # ----------------------------------------------------------
    data_orig, times = archivo.get_data(picks=canales, return_times=True)

    # ----------------------------------------------------------
    # ETAPA 1 — FILTRO PASA-BANDA (1–40 Hz)
    # ----------------------------------------------------------
    # Se trabaja siempre sobre copias (.copy()) para no modificar
    # el objeto original 'archivo'. Esto permite comparar etapas.
    #
    # l_freq=1.0 Hz (límite inferior):
    #   Elimina la deriva de línea de base (componentes DC y
    #   oscilaciones muy lentas < 1 Hz causadas por movimiento
    #   del electrodo o sudoración).
    #
    # h_freq=40.0 Hz (límite superior):
    #   Conserva las bandas de interés:
    #     delta (0.5–4 Hz), theta (4–8 Hz),
    #     mu/alfa (8–13 Hz), beta (13–30 Hz)
    #   Elimina ruido muscular (EMG) que aparece por encima de 40 Hz.
    # ----------------------------------------------------------
    archivo_paso1 = archivo.copy().filter(l_freq=1.0, h_freq=40.0, verbose=False)
    data_paso1, _ = archivo_paso1.get_data(picks=canales, return_times=True)

    # ----------------------------------------------------------
    # ETAPA 2 — FILTRO NOTCH (60 Hz)
    # ----------------------------------------------------------
    # La red eléctrica en Colombia (y en EE.UU.) opera a 60 Hz.
    # Esta frecuencia se acopla electromagnéticamente al amplificador
    # EEG, creando un pico muy pronunciado en el espectro.
    #
    # El filtro notch elimina selectivamente esa frecuencia sin
    # afectar las frecuencias vecinas (59 Hz y 61 Hz quedan intactas),
    # a diferencia de un filtro pasa-banda que cortaría todo.
    #
    # Se aplica DESPUÉS del pasa-banda para no interferir con él.
    # ----------------------------------------------------------
    archivo_paso2 = archivo_paso1.copy().notch_filter(freqs=60.0, verbose=False)
    data_paso2, _ = archivo_paso2.get_data(picks=canales, return_times=True)

    # ----------------------------------------------------------
    # ETAPA 3 — REREFERENCIACIÓN (Common Average Reference - CAR)
    # ----------------------------------------------------------
    # Toda señal EEG es relativa a una referencia. Si esa referencia
    # tiene ruido, contamina todos los canales.
    #
    # La CAR calcula el promedio de TODOS los electrodos en cada
    # instante de tiempo y lo resta a cada canal:
    #   V_car(t) = V_canal(t) - mean(V_todos_los_canales(t))
    #
    # Ventajas:
    #   - Elimina ruido que es común a todos los electrodos
    #     (artefactos ambientales, movimiento del sujeto)
    #   - Mejora la resolución espacial de la señal
    #   - Es el estándar recomendado para análisis de imaginería motora
    # ----------------------------------------------------------
    archivo_proc = archivo_paso2.copy().set_eeg_reference('average', verbose=False)
    data_proc, _ = archivo_proc.get_data(picks=canales, return_times=True)

    # ----------------------------------------------------------
    # VISUALIZACIÓN
    # ----------------------------------------------------------
    if graficar:

        # Máscara de tiempo: solo graficar los primeros t_max_plot segundos
        mascara = times <= t_max_plot
        t_plot  = times[mascara]
        UV      = 1e6   # conversión V → µV

        # Paleta de colores
        colores_canal  = ['#1f77b4', '#d62728', '#2ca02c']   # C3=azul, C4=rojo, Cz=verde
        colores_evento = {'T0': '#d0e8ff', 'T1': '#ffe0b2', 'T2': '#c8f0c8'}
        nombres_evento = {'T0': 'Reposo', 'T1': 'Imag. izquierda', 'T2': 'Imag. derecha'}

        # ---- Función auxiliar interna ----
        # Dibuja las franjas de color de los eventos en un eje dado.
        # Se define aquí dentro para que capture 'archivo', 'colores_evento'
        # y 'nombres_evento' del ámbito de procesar_eeg() directamente.
        def _dibujar_eventos(ax, t_max):
            """Dibuja axvspan coloreados por tipo de evento T0/T1/T2."""
            primeras = set()
            for ann in raw.annotations:
                onset = ann['onset']
                dur   = ann['duration']
                desc  = ann['description']
                if onset > t_max:
                    continue
                color = colores_evento.get(desc, '#eeeeee')
                label = nombres_evento.get(desc, desc) if desc not in primeras else None
                ax.axvspan(
                    onset, min(onset + dur, t_max),
                    alpha=0.18, color=color, label=label, zorder=0
                )
                primeras.add(desc)

        # ===========================================================
        # FIGURA 1 — Las 4 etapas, una fila por etapa
        # ===========================================================
        # Permite ver visualmente el efecto acumulado de cada filtro.
        # Los 3 canales se superponen en cada fila con colores distintos.
        # ===========================================================

        etapas = [
            (data_orig,  'Etapa 0 — Señal original (sin filtrar)'),
            (data_paso1, 'Etapa 1 — Filtro pasa-banda 1–40 Hz  '
                         '(se elimina deriva de línea base y ruido muscular)'),
            (data_paso2, 'Etapa 2 — Filtro Notch 60 Hz  '
                         '(se elimina interferencia de la red eléctrica)'),
            (data_proc,  'Etapa 3 — Rereferenciación CAR  '
                         '(se reduce el ruido común a todos los electrodos)'),
        ]

        fig1, axes1 = plt.subplots(
            4, 1,
            figsize=(16, 14),
            sharex=True,          # todos los subplots comparten el eje X
            facecolor='#f8f9fa'
        )
        fig1.suptitle(
            f'Figura 1 — Efecto del preprocesamiento por etapas\n'
            f'Canales C3, C4, Cz | Primeros {t_max_plot:.0f} s',
            fontsize=13, fontweight='bold', y=1.01
        )

        for i, (data_etapa, titulo) in enumerate(etapas):
            ax = axes1[i]
            ax.set_facecolor('white')
            _dibujar_eventos(ax, t_max=t_max_plot)

            for j, ch in enumerate(canales):
                ax.plot(
                    t_plot,
                    data_etapa[j, mascara] * UV,
                    color=colores_canal[j],
                    linewidth=0.75,
                    label=ch,
                    zorder=2
                )

            ax.set_title(titulo, fontsize=10, fontweight='bold', pad=4)
            ax.set_ylabel('Amplitud (µV)', fontsize=9)
            ax.grid(True, alpha=0.25, linestyle='--')

            # Leyenda: canales a la derecha, eventos a la izquierda
            handles, labels = ax.get_legend_handles_labels()
            ax.legend(handles, labels, loc='upper right',
                      fontsize=8, framealpha=0.85, ncol=4)

        axes1[-1].set_xlabel('Tiempo (s)', fontsize=10)
        plt.tight_layout()
        plt.savefig('fig1_etapas_preprocesamiento.png',
                    dpi=150, bbox_inches='tight')
        plt.show()
        plt.close()
        print("✓ Figura 1 guardada — etapas de preprocesamiento")

        # ===========================================================
        # FIGURA 2 — Original vs Procesada, una fila por canal
        # ===========================================================
        # Muestra de forma directa y clara cuánto cambia la señal
        # de cada canal al pasar por el pipeline completo.
        # La señal original en gris, la procesada en color.
        # ===========================================================

        fig2, axes2 = plt.subplots(
            len(canales), 1,
            figsize=(16, 4 * len(canales)),
            sharex=True,
            facecolor='#f8f9fa'
        )
        # Si solo hay un canal, axes2 no es lista → convertir
        if len(canales) == 1:
            axes2 = [axes2]

        fig2.suptitle(
            f'Figura 2 — Señal original vs procesada por canal\n'
            f'Primeros {t_max_plot:.0f} s',
            fontsize=13, fontweight='bold', y=1.01
        )

        for j, ch in enumerate(canales):
            ax = axes2[j]
            ax.set_facecolor('white')
            _dibujar_eventos(ax, t_max=t_max_plot)

            # Señal original en gris semitransparente (fondo)
            ax.plot(
                t_plot,
                data_orig[j, mascara] * UV,
                color='#aaaaaa',
                linewidth=0.6,
                alpha=0.75,
                label='Original',
                zorder=2
            )
            # Señal procesada en color (primer plano)
            ax.plot(
                t_plot,
                data_proc[j, mascara] * UV,
                color=colores_canal[j],
                linewidth=0.9,
                label='Procesada (pasa-banda + notch + CAR)',
                zorder=3
            )

            ax.set_title(f'Canal {ch}', fontsize=11,
                         fontweight='bold', pad=4)
            ax.set_ylabel('Amplitud (µV)', fontsize=9)
            ax.legend(loc='upper right', fontsize=8, framealpha=0.85)
            ax.grid(True, alpha=0.25, linestyle='--')

        axes2[-1].set_xlabel('Tiempo (s)', fontsize=10)
        plt.tight_layout()
        plt.savefig('fig2_original_vs_procesada.png',
                    dpi=150, bbox_inches='tight')
        plt.show()
        plt.close()
        print("✓ Figura 2 guardada — original vs procesada por canal")

        # ===========================================================
        # FIGURA 3 — PSD antes vs después del procesamiento
        # ===========================================================
        # La Densidad Espectral de Potencia (PSD) muestra cuánta
        # energía tiene la señal en cada frecuencia.
        #
        # Esta figura es clave para VERIFICAR que:
        #   1. El filtro pasa-banda cortó < 1 Hz y > 40 Hz
        #   2. El filtro notch eliminó el pico de 60 Hz
        #
        # Se usa el método de Welch: divide la señal en ventanas
        # solapadas, calcula el espectro de cada una y promedia.
        # Esto reduce la varianza de la estimación espectral.
        # ===========================================================

        fig3, axes3 = plt.subplots(
            1, 2,
            figsize=(15, 5),
            facecolor='#f8f9fa'
        )
        fig3.suptitle(
            'Figura 3 — Densidad Espectral de Potencia (PSD)\n'
            'Antes vs Después del preprocesamiento',
            fontsize=13, fontweight='bold'
        )

        # PSD de la señal ORIGINAL (1–80 Hz para ver el pico de 60 Hz)
        archivo.compute_psd(
            method='welch',
            fmin=1, fmax=80,
            picks=canales,
            verbose=False
        ).plot(axes=axes3[0], show=False)
        axes3[0].set_title(
            'PSD — Señal original\n'
            '(pico de 60 Hz visible)',
            fontsize=10, fontweight='bold'
        )
        axes3[0].set_facecolor('white')

        # PSD de la señal PROCESADA
        archivo_proc.compute_psd(
            method='welch',
            fmin=1, fmax=80,
            picks=canales,
            verbose=False
        ).plot(axes=axes3[1], show=False)
        axes3[1].set_title(
            'PSD — Señal procesada\n'
            '(pico de 60 Hz eliminado, espectro limitado a 40 Hz)',
            fontsize=10, fontweight='bold'
        )
        axes3[1].set_facecolor('white')

        plt.tight_layout()
        plt.savefig('fig3_psd_antes_despues.png',
                    dpi=150, bbox_inches='tight')
        plt.show()
        plt.close()
        print("✓ Figura 3 guardada — PSD antes vs después")

    return archivo_proc
````

**3.2 Cálculo de la densidad espectral de potencia en conjunto de canales de interés para las diferentes bandas en las cuales se divide el EEG**

````python
# ============================================================
# CÁLCULO DE LA DENSIDAD ESPECTRAL DE POTENCIA
# ============================================================
# Recibe una señal procesada y devuelve la PSD completa para
# los canales de interés. Se separa del procesamiento para
# poder reutilizarla de forma modular en el pipeline masivo.
#
# ¿Por qué Welch?
#   El método de Welch divide la señal en ventanas solapadas,
#   calcula el periodograma de cada ventana y promedia.
#   Ventaja: reduce la varianza de la estimación espectral
#   a costa de una ligera reducción de resolución frecuencial.
#   Es el estándar en análisis EEG por su balance entre
#   resolución y estabilidad.
#
# ¿Por qué n_fft=256?
#   Con sfreq=160 Hz → resolución = 160/256 ≈ 0.625 Hz/bin.
#   Suficiente para separar delta (0.5–4), theta (4–8),
#   mu (8–13) y beta (13–30) sin sobreajustar el espectro.
# ============================================================

def calcular_psd(archivo_proc, canales=None, fmin=1.0, fmax=40.0):
    """
    Calcula la Densidad Espectral de Potencia (PSD) usando
    el método de Welch sobre los canales especificados.

    Parámetros
    ----------
    archivo_proc : mne.io.Raw
        Señal EEG preprocesada (salida de procesar_eeg).
    canales : list, opcional
        Lista de canales a analizar.
        Por defecto ['C3', 'C4', 'Cz'].
    fmin : float
        Frecuencia mínima del análisis espectral (Hz).
    fmax : float
        Frecuencia máxima del análisis espectral (Hz).

    Retorna
    -------
    psd : np.ndarray, shape (n_canales, n_frecuencias)
        Densidad espectral en µV²/Hz.
        Cada fila corresponde a un canal, cada columna a
        una frecuencia del vector 'freqs'.
    freqs : np.ndarray, shape (n_frecuencias,)
        Vector de frecuencias en Hz correspondiente a
        las columnas de 'psd'.
    canales_usados : list
        Nombres de los canales efectivamente analizados,
        en el mismo orden que las filas de 'psd'.
    """

    # Validar y filtrar canales disponibles
    # Si el archivo no tiene algún canal pedido, se omite
    # en lugar de lanzar un error que detenga el pipeline.
    if canales is None:
        canales = ['C3', 'C4', 'Cz']

    canales_usados = [ch for ch in canales if ch in archivo_proc.ch_names]

    if not canales_usados:
        raise ValueError(
            f"Ninguno de los canales {canales} está disponible "
            f"en el archivo. Canales presentes: {archivo_proc.ch_names}"
        )

    # --- Extraer datos como array numpy ---
    # archivo_proc.get_data() → (n_canales, n_muestras) en voltios
    # Se trabaja directamente con numpy para máxima eficiencia
    # vectorial (sin bucles sobre canales o muestras).
    data = archivo_proc.get_data(picks=canales_usados)  # V

    sfreq = archivo_proc.info['sfreq']   # frecuencia de muestreo en Hz

    # --- Calcular PSD con Welch vectorizado ---
    # psd_array_welch opera sobre (n_canales, n_muestras)
    # simultáneamente → un solo llamado para todos los canales.
    #
    # n_fft=256    : tamaño de la ventana FFT
    # n_overlap=128: solapamiento del 50% entre ventanas (estándar)
    # window='hamming': ventana de Hamming → reduce el efecto
    #                   de fuga espectral en los bordes de cada
    #                   segmento.
    psd_v2, freqs = mne.time_frequency.psd_array_welch(
        data,
        sfreq=sfreq,
        fmin=fmin,
        fmax=fmax,
        n_fft=256,
        n_overlap=128,
        window='hamming',
        verbose=False
    )
    # psd_v2 → (n_canales, n_frecuencias) en V²/Hz

    # Convertir V²/Hz → µV²/Hz para escala clínica estándar
    # 1 V² = 1e12 µV²
    psd_uv2 = psd_v2 * 1e12

    return psd_uv2, freqs, canales_usados


# ============================================================
# EXTRACCIÓN DE POTENCIA POR BANDAS
# ============================================================
# Toma la PSD completa y calcula un único valor de potencia
# representativo por banda y por canal, integrando (promediando)
# la PSD dentro del rango de frecuencias de cada banda.
#
# ¿Por qué promediar en lugar de sumar?
#   Al promediar, el resultado es independiente del número de
#   bins de frecuencia que caen en cada banda (que varía según
#   n_fft y el ancho de la banda). Esto hace los valores
#   comparables entre bandas de distinto ancho.
# ============================================================

def potencia_bandas(psd, freqs, canales_usados):
    """
    Extrae la potencia media en cada banda EEG de interés.

    Las bandas analizadas son:
        delta : 0.5 –  4 Hz  (no siempre relevante post-filtro)
        theta :  4  –  8 Hz
        mu    :  8  – 13 Hz  ← banda principal de las hipótesis
        beta  : 13  – 30 Hz  ← banda secundaria de las hipótesis

    Parámetros
    ----------
    psd : np.ndarray, shape (n_canales, n_frecuencias)
        PSD en µV²/Hz (salida de calcular_psd).
    freqs : np.ndarray
        Vector de frecuencias en Hz (salida de calcular_psd).
    canales_usados : list
        Nombres de los canales en el mismo orden que las
        filas de psd (salida de calcular_psd).

    Retorna
    -------
    resultados : dict
        Diccionario plano con claves del tipo
        'potencia_mu_C3', 'potencia_beta_C4', etc.
        Listo para construir una fila del DataFrame.
    """

    # Definición de bandas como diccionario ordenado
    # Clave → nombre de la banda
    # Valor → (frecuencia mínima, frecuencia máxima) en Hz
    bandas = {
        'delta': (0.5,  4.0),
        'theta': (4.0,  8.0),
        'mu'   : (8.0, 13.0),
        'beta' : (13.0, 30.0),
    }

    resultados = {}

    for nombre_banda, (fmin_b, fmax_b) in bandas.items():

        # Máscara booleana: True en los índices de frecuencia
        # que caen dentro del rango [fmin_b, fmax_b].
        # np.logical_and es equivalente a (freqs >= fmin_b) & (freqs <= fmax_b)
        # pero más explícito y ligeramente más rápido.
        idx_banda = np.logical_and(freqs >= fmin_b, freqs <= fmax_b)

        # Verificar que al menos un bin de frecuencia cae en la banda.
        # Con fmin=1 Hz en calcular_psd, delta podría tener pocos bins.
        if not np.any(idx_banda):
            # Si no hay bins, asignar NaN para no contaminar estadísticas
            for ch in canales_usados:
                resultados[f'potencia_{nombre_banda}_{ch}'] = np.nan
            continue

        # psd[:, idx_banda] → submatriz (n_canales, n_bins_en_banda)
        # .mean(axis=1)     → promedio sobre frecuencias → (n_canales,)
        # Operación vectorizada: no hay bucle sobre canales.
        potencia_por_canal = psd[:, idx_banda].mean(axis=1)

        # Guardar un valor por canal con nombre descriptivo
        # Ejemplo: 'potencia_mu_C3', 'potencia_mu_C4', 'potencia_mu_Cz'
        for i, ch in enumerate(canales_usados):
            resultados[f'potencia_{nombre_banda}_{ch}'] = potencia_por_canal[i]

    return resultados
````

````python
# ============================================================
# PRUEBA DE calcular_psd() Y potencia_bandas()
# ============================================================
# Se usan archivo_procesado y los canales C3, C4, Cz del sujeto 1.
# Al final se imprime una tabla resumen con los valores
# de potencia por banda y canal.
# ============================================================

# Paso 1 — Calcular PSD
psd, freqs, canales_usados = calcular_psd(archivo_procesado)

print(f"PSD calculada:")
print(f"  Shape         : {psd.shape}  → ({len(canales_usados)} canales × {len(freqs)} frecuencias)")
print(f"  Canales       : {canales_usados}")
print(f"  Rango freqs   : {freqs[0]:.2f} – {freqs[-1]:.2f} Hz")
print(f"  Resolución    : {freqs[1]-freqs[0]:.3f} Hz/bin")
print()

# Paso 2 — Extraer potencia por bandas
pot = potencia_bandas(psd, freqs, canales_usados)

# Mostrar como tabla organizada
print(f"{'Banda':<8} {'Canal':<6} {'Potencia (µV²/Hz)':>20}")
print("-" * 38)
for clave, valor in pot.items():
    _, banda, canal = clave.split('_', 2)
    print(f"{banda:<8} {canal:<6} {valor:>20.4f}")
````

**3.3 Rutina que aplica sobre todos los archivos de la base de datos las rutinas anteriores y almacena los resultados en un dataframe donde se puede discriminar nombre sujeto, la condición evaluada y la densidad espectral de potencia obtenida**

````python
# ============================================================
# MAPEO DE RUNS → CONDICIONES EXPERIMENTALES
# ============================================================
# El dataset PhysioNet organiza los registros así:
#
#   Run 1        → baseline ojos abiertos  (no se usa)
#   Run 2        → baseline ojos cerrados  (no se usa)
#   Run 3,7,11   → movimiento REAL         (T1=izq, T2=der)
#   Run 4,8,12   → imaginación motora      (T1=izq, T2=der)
#   Run 5,9,13   → movimiento REAL         (T1=ambas, T2=ambas) → no aplica
#   Run 6,10,14  → imaginación motora      (T1=ambas, T2=ambas) → no aplica
#
# Para este proyecto interesa:
#   MM (movimiento real)   → runs 3, 7, 11
#   MI (imaginación)       → runs 4, 8, 12
#
# Dentro de cada run, las anotaciones indican la condición:
#   T0 → reposo         (entre tareas)
#   T1 → mano izquierda
#   T2 → mano derecha
#
# La condición completa se construye combinando tipo_run + T:
#   run MM + T1 → 'MM_L'  (movimiento real mano izquierda)
#   run MM + T2 → 'MM_R'  (movimiento real mano derecha)
#   run MI + T1 → 'MI_L'  (imaginación mano izquierda)
#   run MI + T2 → 'MI_R'  (imaginación mano derecha)
#   cualquier  + T0 → 'REST'
# ============================================================

# Diccionario: run → tipo de tarea
RUNS_TAREA = {
    3:  'MM',   # movimiento real mano izq/der
    7:  'MM',
    11: 'MM',
    4:  'MI',   # imaginación mano izq/der
    8:  'MI',
    12: 'MI',
}

# Función que traduce (tipo_run, anotación) → etiqueta de condición
def get_condicion(tipo_run, anotacion):
    """
    Retorna la etiqueta de condición para una combinación
    de tipo de run y anotación de evento.

    Parámetros
    ----------
    tipo_run : str
        'MM' (movimiento real) o 'MI' (imaginación motora).
    anotacion : str
        'T0', 'T1' o 'T2'.

    Retorna
    -------
    str : 'REST', 'MM_L', 'MM_R', 'MI_L' o 'MI_R'
    """
    if anotacion == 'T0':
        return 'REST'
    elif anotacion == 'T1':
        return f'{tipo_run}_L'   # mano izquierda
    elif anotacion == 'T2':
        return f'{tipo_run}_R'   # mano derecha
    else:
        return 'UNKNOWN'


# Verificación rápida del mapeo
print("Verificación del mapeo de condiciones:")
print(f"  Run 4  (MI) + T0 → {get_condicion('MI', 'T0')}")
print(f"  Run 4  (MI) + T1 → {get_condicion('MI', 'T1')}")
print(f"  Run 4  (MI) + T2 → {get_condicion('MI', 'T2')}")
print(f"  Run 3  (MM) + T1 → {get_condicion('MM', 'T1')}")
print(f"  Run 11 (MM) + T2 → {get_condicion('MM', 'T2')}")
````

````python
# ============================================================
# EXTRAER ÉPOCAS Y PSD POR CONDICIÓN
# ============================================================
# Dado un archivo raw procesado, identifica cada evento
# (T0/T1/T2), extrae el segmento de señal correspondiente
# y calcula la PSD + potencia por bandas para ese segmento.
#
# ¿Por qué segmentar por evento antes de calcular la PSD?
#   Si se calcula la PSD sobre toda la señal continua,
#   se mezclan todas las condiciones y se pierde la capacidad
#   de comparar MI_R vs MI_L vs REST.
#   Al segmentar por evento, cada PSD corresponde exactamente
#   a una tarea específica.
# ============================================================

def extraer_epocas_psd(raw_proc, tipo_run, sujeto_id,
                       canales=None, duracion_epoca=4.0):
    """
    Extrae segmentos de señal por evento y calcula la PSD
    de cada uno, devolviendo una lista de filas para el DataFrame.

    Parámetros
    ----------
    raw_proc : mne.io.Raw
        Señal preprocesada (salida de procesar_eeg).
    tipo_run : str
        'MM' o 'MI' según el run actual.
    sujeto_id : str
        Identificador del sujeto, p.ej. 'S001'.
    canales : list, opcional
        Canales a analizar. Por defecto ['C3', 'C4', 'Cz'].
    duracion_epoca : float
        Duración en segundos de cada época a extraer.
        Por defecto 4 s (duración estándar de cada tarea
        en el protocolo PhysioNet EEGBCI).

    Retorna
    -------
    filas : list of dict
        Cada elemento es un diccionario con:
            'sujeto'    → str, p.ej. 'S001'
            'run_tipo'  → str, 'MM' o 'MI'
            'condicion' → str, 'REST'/'MM_L'/'MM_R'/'MI_L'/'MI_R'
            'epoca'     → int, índice de la época (0, 1, 2...)
            'potencia_<banda>_<canal>' → float, µV²/Hz
    """

    if canales is None:
        canales = ['C3', 'C4', 'Cz']

    canales_usados = [ch for ch in canales if ch in raw_proc.ch_names]
    sfreq          = raw_proc.info['sfreq']

    # n_muestras por época: duración × frecuencia de muestreo
    # int() trunca para no exceder el buffer de la señal
    n_muestras_epoca = int(duracion_epoca * sfreq)

    filas = []

    # Contador de épocas por condición (para indexarlas)
    contadores = {}

    for ann in raw_proc.annotations:
        onset    = ann['onset']      # inicio del evento en segundos
        desc     = ann['description']  # 'T0', 'T1' o 'T2'

        # Traducir anotación a etiqueta de condición
        condicion = get_condicion(tipo_run, desc)
        if condicion == 'UNKNOWN':
            continue

        # Inicio y fin del segmento en muestras
        idx_inicio = int(onset * sfreq)
        idx_fin    = idx_inicio + n_muestras_epoca

        # Verificar que el segmento no excede la señal
        n_total = raw_proc.n_times
        if idx_fin > n_total:
            continue   # segmento incompleto → descartar

        # --- Extraer segmento de señal ---
        # raw_proc.get_data() acepta start/stop en muestras.
        # Resultado: (n_canales, n_muestras_epoca) en voltios
        segmento = raw_proc.get_data(
            picks=canales_usados,
            start=idx_inicio,
            stop=idx_fin
        )

        # --- Calcular PSD del segmento con Welch ---
        # Mismos parámetros que calcular_psd() para consistencia.
        # n_fft mínimo: no puede ser mayor que n_muestras_epoca.
        n_fft_seg = min(256, n_muestras_epoca)

        psd_v2, freqs = mne.time_frequency.psd_array_welch(
            segmento,
            sfreq=sfreq,
            fmin=1.0,
            fmax=40.0,
            n_fft=n_fft_seg,
            n_overlap=n_fft_seg // 2,
            window='hamming',
            verbose=False
        )

        # Convertir V²/Hz → µV²/Hz
        psd_uv2 = psd_v2 * 1e12

        # --- Extraer potencia por bandas ---
        pot = potencia_bandas(psd_uv2, freqs, canales_usados)

        # --- Construir fila del DataFrame ---
        # Índice de época para esta condición en este run
        contadores[condicion] = contadores.get(condicion, 0)
        fila = {
            'sujeto'   : sujeto_id,
            'run_tipo' : tipo_run,
            'condicion': condicion,
            'epoca'    : contadores[condicion],
        }
        fila.update(pot)   # agrega potencia_mu_C3, potencia_beta_C4, etc.

        filas.append(fila)
        contadores[condicion] += 1

    return filas
````

````python
RUTA_DATOS = r'C:\Users\mrios\Desktop\Universidad\Semestre VII\Bioseñales\files'

def get_ruta_local(sujeto, run):
    """
    Construye la ruta local al archivo EDF dado un sujeto y run.

    Parámetros
    ----------
    sujeto : int
        Número de sujeto (1 a 109).
    run : int
        Número de run (3, 4, 7, 8, 11, 12).

    Retorna
    -------
    str : ruta absoluta al archivo .edf
    """
    sujeto_str  = f'S{sujeto:03d}'          # 1 → 'S001'
    archivo_str = f'S{sujeto:03d}R{run:02d}.edf'  # 1,4 → 'S001R04.edf'
    return os.path.join(RUTA_DATOS, sujeto_str, archivo_str)


# ============================================================
# PIPELINE MASIVO 
# ============================================================

def pipeline_masivo(n_sujetos=109, canales=None,
                    duracion_epoca=4.0, verbose=True):
    """
    Aplica el pipeline completo sobre todos los sujetos
    leyendo los archivos desde una carpeta local.

    Parámetros
    ----------
    n_sujetos : int
        Número de sujetos a procesar (1 a 109).
        Usar n_sujetos=3 para pruebas rápidas.
    canales : list, opcional
        Canales a analizar. Por defecto ['C3', 'C4', 'Cz'].
    duracion_epoca : float
        Duración de cada época en segundos (por defecto 4 s).
    verbose : bool
        Si True, imprime progreso por sujeto.

    Retorna
    -------
    df : pd.DataFrame
        Una fila por época × condición × run × sujeto.
    """

    if canales is None:
        canales = ['C3', 'C4', 'Cz']

    # Acumulador de filas — más eficiente que pd.concat() iterativo
    todas_las_filas = []

    # Contadores para el resumen final
    sujetos_ok    = 0
    sujetos_error = 0
    archivos_ok   = 0

    print("=" * 60)
    print(f"INICIANDO PIPELINE — {n_sujetos} sujetos (archivos locales)")
    print(f"Ruta base  : {RUTA_DATOS}")
    print(f"Runs       : {list(RUNS_TAREA.keys())}")
    print(f"Canales    : {canales}")
    print("=" * 60)

    for sujeto in range(1, n_sujetos + 1):

        sujeto_id    = f'S{sujeto:03d}'
        filas_sujeto = []
        errores_sujeto = 0

        for run, tipo_run in RUNS_TAREA.items():
            try:
                # --- Construir ruta local ---
                ruta = get_ruta_local(sujeto, run)

                # Verificar que el archivo existe antes de intentar cargarlo
                # Así el mensaje de error es claro: "archivo no encontrado"
                # en lugar de un error críptico de MNE.
                if not os.path.exists(ruta):
                    raise FileNotFoundError(
                        f"Archivo no encontrado: {ruta}"
                    )

                # --- Cargar ---
                raw = cargar_eeg_con_eventos(ruta)

                # --- Preprocesar SIN graficar ---
                # graficar=False es crítico para el pipeline masivo:
                # evita generar 654 figuras y consume mucho menos memoria.
                raw_proc = procesar_eeg(raw, graficar=False)

                # --- Extraer épocas y PSD por condición ---
                filas = extraer_epocas_psd(
                    raw_proc,
                    tipo_run,
                    sujeto_id,
                    canales=canales,
                    duracion_epoca=duracion_epoca
                )

                filas_sujeto.extend(filas)
                archivos_ok += 1

            except FileNotFoundError as e:
                # Error esperado si faltan archivos locales
                errores_sujeto += 1
                if verbose:
                    print(f"  ✗ {sujeto_id} Run {run:02d} — {e}")

            except Exception as e:
                # Cualquier otro error (archivo corrupto, canal faltante, etc.)
                errores_sujeto += 1
                if verbose:
                    print(f"  ✗ {sujeto_id} Run {run:02d} — Error inesperado: {e}")

        # Solo contar el sujeto si al menos un run fue exitoso
        if filas_sujeto:
            todas_las_filas.extend(filas_sujeto)
            sujetos_ok += 1
            if verbose:
                n_epocas = len(filas_sujeto)
                print(f"  ✓ {sujeto_id} — {n_epocas} épocas  "
                      f"| {6 - errores_sujeto}/6 runs exitosos")
        else:
            sujetos_error += 1
            if verbose:
                print(f"  ✗ {sujeto_id} — sin datos válidos")

    # --- Construir DataFrame de una sola vez ---
    # Crear el DataFrame al final (no en cada iteración) es
    # la forma más eficiente en memoria y tiempo.
    df = pd.DataFrame(todas_las_filas)

    if df.empty:
        print("\n⚠ El DataFrame está vacío. Verifica la ruta y estructura de archivos.")
        return df

    # Columnas categóricas → menos memoria, groupby más rápido
    for col in ['sujeto', 'run_tipo', 'condicion']:
        if col in df.columns:
            df[col] = df[col].astype('category')

    # --- Resumen final ---
    print("\n" + "=" * 60)
    print("PIPELINE COMPLETADO")
    print(f"  Sujetos procesados  : {sujetos_ok} / {n_sujetos}")
    print(f"  Sujetos con error   : {sujetos_error}")
    print(f"  Archivos procesados : {archivos_ok} / {n_sujetos * 6}")
    print(f"  Total de épocas     : {len(df)}")
    print(f"  Shape del DataFrame : {df.shape}")
    print(f"  Condiciones únicas  : {sorted(df['condicion'].unique())}")
    print(f"  Columnas            : {list(df.columns)}")
    print("=" * 60)

    return df
````

Antes de aplicar pruebas estadísticas, se exploran los datos visualmente para identificar patrones, diferencias entre condiciones y posibles valores atípicos. 


````python
# FIGURAS EN ESTE PASO:
#   Fig 5 — Boxplots de potencia mu y beta por condición
#   Fig 6 — Boxplots de lateralización C3 vs C4 por condición
#   Fig 7 — Tabla de estadísticos descriptivos (media ± DE)
#   Fig 8 — Mapa de calor de potencia media por condición y canal

# Paleta de colores por condición — consistente en todas las figuras
COLORES_CONDICION = {
    'REST' : '#aaaaaa',
    'MI_L' : '#1f77b4',
    'MI_R' : '#d62728',
    'MM_L' : '#2ca02c',
    'MM_R' : '#ff7f0e',
}

# Orden fijo de condiciones para todos los gráficos
ORDEN_CONDICIONES = ['REST', 'MI_L', 'MI_R', 'MM_L', 'MM_R']

# Etiquetas legibles para los ejes
ETIQUETAS_CONDICION = {
    'REST' : 'Reposo',
    'MI_L' : 'Imag. izq.\n(MI_L)',
    'MI_R' : 'Imag. der.\n(MI_R)',
    'MM_L' : 'Mov. real izq.\n(MM_L)',
    'MM_R' : 'Mov. real der.\n(MM_R)',
}
````

````python
# ============================================================
# BOXPLOTS DE POTENCIA MU Y BETA POR CONDICIÓN
# ============================================================
# Un boxplot muestra la mediana, el rango intercuartílico (IQR)
# y los valores atípicos de una distribución.
# Es la representación más adecuada para comparar distribuciones
# de potencia EEG porque:
#   - Los datos EEG no son normales (sesgados a la derecha)
#   - La mediana es más robusta que la media ante outliers
#   - Permite ver la dispersión de cada condición
#
# Se grafica mu y beta por separado para C3 y C4, que son los
# canales directamente relacionados con las hipótesis.
# ============================================================

def figura_boxplots_bandas(df):
    """
    Genera boxplots de potencia mu y beta en C3 y C4
    para cada condición experimental.
    """

    # Bandas y canales a visualizar
    # Solo mu y beta porque son los relevantes para las hipótesis
    configs = [
        ('potencia_mu_C3',   'Banda mu (8–13 Hz) — Canal C3',  '#1f77b4'),
        ('potencia_mu_C4',   'Banda mu (8–13 Hz) — Canal C4',  '#d62728'),
        ('potencia_beta_C3', 'Banda beta (13–30 Hz) — Canal C3', '#1f77b4'),
        ('potencia_beta_C4', 'Banda beta (13–30 Hz) — Canal C4', '#d62728'),
    ]

    fig, axes = plt.subplots(2, 2, figsize=(16, 11), facecolor='#f8f9fa')
    fig.suptitle(
        'Figura 5 — Potencia espectral por condición\n'
        'Bandas mu (8–13 Hz) y beta (13–30 Hz) | Canales C3 y C4',
        fontsize=13, fontweight='bold', y=1.01
    )

    axes = axes.flatten()   # convertir matriz 2×2 en lista de 4 ejes

    for idx, (columna, titulo, color_canal) in enumerate(configs):
        ax = axes[idx]
        ax.set_facecolor('white')

        # Preparar datos por condición en el orden definido
        # Solo incluir condiciones que existen en el DataFrame
        condiciones_presentes = [
            c for c in ORDEN_CONDICIONES if c in df['condicion'].values
        ]

        datos_por_condicion = [
            df[df['condicion'] == cond][columna].dropna().values
            for cond in condiciones_presentes
        ]

        # --- Boxplot base ---
        bp = ax.boxplot(
            datos_por_condicion,
            patch_artist=True,      # relleno de color en las cajas
            notch=False,            # sin muesca (notch requiere más datos)
            showfliers=True,        # mostrar outliers como puntos
            flierprops=dict(
                marker='o',
                markersize=2.5,
                alpha=0.3,
                linestyle='none'
            ),
            medianprops=dict(color='black', linewidth=2),
            whiskerprops=dict(linewidth=1.2),
            capprops=dict(linewidth=1.2),
        )

        # Colorear cada caja según la condición
        for patch, cond in zip(bp['boxes'], condiciones_presentes):
            patch.set_facecolor(COLORES_CONDICION[cond])
            patch.set_alpha(0.65)

        # Superponer puntos individuales (strip plot manual)
        # Permite ver la distribución real de los datos
        # np.random.seed garantiza reproducibilidad del jitter
        np.random.seed(42)
        for i, (datos, cond) in enumerate(
            zip(datos_por_condicion, condiciones_presentes)
        ):
            # Submuestreo: si hay muchos puntos, mostrar solo 80
            # para no saturar la figura
            if len(datos) > 80:
                indices = np.random.choice(len(datos), 80, replace=False)
                datos_plot = datos[indices]
            else:
                datos_plot = datos

            # Jitter horizontal para separar los puntos
            jitter = np.random.uniform(-0.2, 0.2, size=len(datos_plot))
            ax.scatter(
                np.full(len(datos_plot), i + 1) + jitter,
                datos_plot,
                color=COLORES_CONDICION[cond],
                alpha=0.25,
                s=8,
                zorder=3
            )

        # Etiquetas del eje X
        etiquetas = [
            ETIQUETAS_CONDICION[c] for c in condiciones_presentes
        ]
        ax.set_xticks(range(1, len(condiciones_presentes) + 1))
        ax.set_xticklabels(etiquetas, fontsize=9)

        ax.set_title(titulo, fontsize=10, fontweight='bold', pad=6)
        ax.set_ylabel('Potencia media (µV²/Hz)', fontsize=9)
        ax.grid(True, axis='y', alpha=0.3, linestyle='--')
        ax.set_xlim(0.4, len(condiciones_presentes) + 0.6)

    plt.tight_layout()
    plt.savefig('fig5_boxplots_bandas.png', dpi=150, bbox_inches='tight')
    plt.show()
    plt.close()
    print("✓ Figura 5 guardada — boxplots mu y beta por condición")


figura_boxplots_bandas(df)
````

````python
# ============================================================
# LATERALIZACIÓN C3 vs C4 POR CONDICIÓN
# ============================================================
# Esta figura es la más directamente relacionada con las
# hipótesis de lateralización.
#
# La idea: si durante MI_R (imaginación mano derecha) hay ERD
# contralateral, entonces la potencia mu en C3 debería ser
# MENOR que en C4 para esa condición.
#
# Se grafica la diferencia C3 - C4 para la banda mu:
#   Valor negativo → C3 < C4 → ERD contralateral (esperado en MI_R)
#   Valor positivo → C3 > C4 → ERD ipsilateral
#   Valor ~0       → sin lateralización
# ============================================================

def figura_lateralizacion(df):
    """
    Grafica la diferencia de potencia mu entre C3 y C4
    (índice de lateralización) por condición.
    """

    fig, axes = plt.subplots(1, 2, figsize=(15, 6), facecolor='#f8f9fa')
    fig.suptitle(
        'Figura 6 — Lateralización hemisférica: diferencia C3 − C4\n'
        '(negativo = mayor activación en C3; '
        'positivo = mayor activación en C4)',
        fontsize=12, fontweight='bold', y=1.02
    )

    for ax, banda, titulo in zip(
        axes,
        ['mu', 'beta'],
        ['Banda mu (8–13 Hz)', 'Banda beta (13–30 Hz)']
    ):
        ax.set_facecolor('white')

        condiciones_presentes = [
            c for c in ORDEN_CONDICIONES if c in df['condicion'].values
        ]

        # Calcular diferencia C3 - C4 por época
        # Una diferencia negativa significa que C3 tiene menos potencia
        # que C4, es decir, hay más activación (ERD) en C3.
        col_c3 = f'potencia_{banda}_C3'
        col_c4 = f'potencia_{banda}_C4'

        datos_dif = []
        for cond in condiciones_presentes:
            subset = df[df['condicion'] == cond]
            diferencia = (subset[col_c3] - subset[col_c4]).dropna().values
            datos_dif.append(diferencia)

        bp = ax.boxplot(
            datos_dif,
            patch_artist=True,
            notch=False,
            showfliers=True,
            flierprops=dict(marker='o', markersize=2, alpha=0.25,
                            linestyle='none'),
            medianprops=dict(color='black', linewidth=2),
            whiskerprops=dict(linewidth=1.2),
            capprops=dict(linewidth=1.2),
        )

        for patch, cond in zip(bp['boxes'], condiciones_presentes):
            patch.set_facecolor(COLORES_CONDICION[cond])
            patch.set_alpha(0.65)

        # Línea horizontal en 0 — referencia: sin lateralización
        ax.axhline(0, color='black', linewidth=1.2,
                   linestyle='--', alpha=0.6, zorder=1)
        ax.text(
            len(condiciones_presentes) + 0.5, 0.3,
            'sin lateralización',
            fontsize=8, color='gray', va='bottom'
        )

        etiquetas = [ETIQUETAS_CONDICION[c] for c in condiciones_presentes]
        ax.set_xticks(range(1, len(condiciones_presentes) + 1))
        ax.set_xticklabels(etiquetas, fontsize=9)
        ax.set_title(titulo, fontsize=10, fontweight='bold', pad=6)
        ax.set_ylabel('Diferencia de potencia C3 − C4 (µV²/Hz)', fontsize=9)
        ax.grid(True, axis='y', alpha=0.3, linestyle='--')
        ax.set_xlim(0.4, len(condiciones_presentes) + 0.6)

    plt.tight_layout()
    plt.savefig('fig6_lateralizacion.png', dpi=150, bbox_inches='tight')
    plt.show()
    plt.close()
    print("✓ Figura 6 guardada — lateralización C3 vs C4")


figura_lateralizacion(df)
````

````python
# ============================================================
# TABLA DE ESTADÍSTICOS DESCRIPTIVOS
# ============================================================

ORDEN_CONDICIONES = ['REST', 'MM_R', 'MM_L', 'MI_R', 'MI_L']

# Columnas
COLUMNAS_INTERES = [
    'potencia_mu_C3',   'potencia_mu_C4',   'potencia_mu_Cz',
    'potencia_beta_C3', 'potencia_beta_C4',  'potencia_beta_Cz',
]

def tabla_descriptivos(df):
    """
    Calcula estadísticos descriptivos (media, DE, mediana, IQR)
    por condición y banda/canal. Usa operaciones vectorizadas de pandas.
    """
    df_desc = (
        df[df['condicion'].isin(ORDEN_CONDICIONES)]
        .groupby('condicion', observed=True)[COLUMNAS_INTERES]
        .agg(['mean', 'std', 'median',
              lambda x: x.quantile(0.75) - x.quantile(0.25)])
        .round(4)
    )

    # Aplanar MultiIndex y renombrar la lambda como IQR
    df_desc.columns = [
        f'{col}_{stat}'.replace('<lambda_0>', 'IQR').replace('<lambda>', 'IQR')
        for col, stat in df_desc.columns
    ]
    df_desc = df_desc.reindex([c for c in ORDEN_CONDICIONES if c in df_desc.index])

    print('=' * 70)
    print('ESTADÍSTICOS DESCRIPTIVOS — Potencia espectral por condición')
    print('=' * 70)

    for col in COLUMNAS_INTERES:
        banda = col.replace('potencia_', '').upper()
        print(f'\n  {banda}')
        print(f'  {"Condición":<12} {"Media":>10} {"± DE":>10} {"Mediana":>10} {"IQR":>10}')
        print('  ' + '-' * 46)
        for cond in ORDEN_CONDICIONES:
            if cond not in df_desc.index:
                continue
            print(f'  {cond:<12} '
                  f'{df_desc.loc[cond, f"{col}_mean"]:>10.4f} '
                  f'{df_desc.loc[cond, f"{col}_std"]:>10.4f} '
                  f'{df_desc.loc[cond, f"{col}_median"]:>10.4f} '
                  f'{df_desc.loc[cond, f"{col}_IQR"]:>10.4f}')

    df_desc.to_csv('descriptivos_eeg.csv')
    print('\n✓ Guardado en descriptivos_eeg.csv')
    return df_desc

df_desc = tabla_descriptivos(df)
````

````python
# ============================================================
# MAPA DE CALOR DE POTENCIA MEDIA
# ============================================================
# Un heatmap muestra de un vistazo cómo varía la potencia
# en cada combinación condición × canal × banda.
# Es una visualización complementaria a los boxplots que
# permite identificar patrones de lateralización globales.
# ============================================================

def figura_heatmap(df):
    """
    Genera un mapa de calor de potencia media normalizada
    por condición y canal para las bandas mu y beta.
    """

    columnas = [
        'potencia_mu_C3',   'potencia_mu_C4',   'potencia_mu_Cz',
        'potencia_beta_C3', 'potencia_beta_C4',  'potencia_beta_Cz',
    ]

    etiquetas_cols = [
        'mu\nC3', 'mu\nC4', 'mu\nCz',
        'beta\nC3', 'beta\nC4', 'beta\nCz',
    ]

    condiciones_presentes = [
        c for c in ORDEN_CONDICIONES if c in df['condicion'].values
    ]

    # Calcular medias por condición — operación vectorizada
    matriz = np.array([
        df[df['condicion'] == cond][columnas].mean().values
        for cond in condiciones_presentes
    ])

    # Normalizar por columna (z-score) para comparar entre bandas
    # Sin normalizar, delta dominaría visualmente por su mayor magnitud.
    # z = (x - media) / std → valores centrados en 0
    matriz_norm = (matriz - matriz.mean(axis=0)) / (matriz.std(axis=0) + 1e-10)

    fig, axes = plt.subplots(1, 2, figsize=(14, 5), facecolor='#f8f9fa')
    fig.suptitle(
        'Figura 8 — Mapa de calor de potencia media por condición y canal\n'
        '(izquierda: valores absolutos µV²/Hz | '
        'derecha: z-score por columna)',
        fontsize=11, fontweight='bold'
    )

    etiquetas_filas = [ETIQUETAS_CONDICION[c].replace('\n', ' ')
                       for c in condiciones_presentes]

    for ax, data, titulo, fmt, cmap in zip(
        axes,
        [matriz, matriz_norm],
        ['Potencia absoluta (µV²/Hz)', 'Potencia normalizada (z-score)'],
        ['.2f', '.2f'],
        ['YlOrRd', 'RdBu_r']
    ):
        ax.set_facecolor('white')

        im = ax.imshow(data, aspect='auto', cmap=cmap)
        plt.colorbar(im, ax=ax, shrink=0.8)

        # Etiquetas de ejes
        ax.set_xticks(range(len(columnas)))
        ax.set_xticklabels(etiquetas_cols, fontsize=9)
        ax.set_yticks(range(len(condiciones_presentes)))
        ax.set_yticklabels(etiquetas_filas, fontsize=9)

        # Valores numéricos dentro de cada celda
        for i in range(len(condiciones_presentes)):
            for j in range(len(columnas)):
                ax.text(
                    j, i, f'{data[i, j]:{fmt}}',
                    ha='center', va='center',
                    fontsize=8,
                    color='white' if abs(data[i, j]) > data.max() * 0.6
                    else 'black'
                )

        ax.set_title(titulo, fontsize=10, fontweight='bold', pad=8)

    plt.tight_layout()
    plt.savefig('fig8_heatmap_potencia.png', dpi=150, bbox_inches='tight')
    plt.show()
    plt.close()
    print("✓ Figura 8 guardada — mapa de calor")


figura_heatmap(df)
````

#### *4.1 Pruebas de hipótesis*

````python
def prueba_hipotesis(df, cond1, cond2, columna, alpha=0.05):
    """
    Prueba si hay diferencia significativa entre dos condiciones
    para una columna de potencia.
    Selecciona automáticamente t-test o Mann-Whitney según normalidad.
    """
    g1 = df[df['condicion'] == cond1][columna].dropna().values
    g2 = df[df['condicion'] == cond2][columna].dropna().values

    print(f'\n{"="*62}')
    print(f'  {cond1} vs {cond2} | {columna.replace("potencia_","").upper()}')
    print(f'  n={len(g1)} vs n={len(g2)}')
    print(f'  Mediana: {np.median(g1):.4f} vs {np.median(g2):.4f}')

    # Shapiro-Wilk — solo válido para n < 5000, tomamos muestra si es necesario
    _, p1 = stats.shapiro(g1[:5000])
    _, p2 = stats.shapiro(g2[:5000])
    normal = (p1 > alpha) and (p2 > alpha)
    print(f'  Shapiro p={p1:.4f}/{p2:.4f} → {"Normal" if normal else "NO normal ← usamos Mann-Whitney"}')

    if normal:
        stat, p = stats.ttest_ind(g1, g2)
        test = 't-test'
    else:
        stat, p = stats.mannwhitneyu(g1, g2, alternative='two-sided')
        test = 'Mann-Whitney U'

    print(f'  Test: {test} | estadístico={stat:.2f} | p={p:.6f}')
    resultado = 'RECHAZA H0 ✓ — diferencia significativa' if p < alpha \
                else 'NO se rechaza H0 — sin diferencia significativa'
    print(f'  → {resultado}')
    return p


# ── H1: Lateralización mu durante MI_R (C3 < C4)
print('\n' + '#'*62)
print('# HIPÓTESIS 1: Mu C3 < Mu C4 durante MI_R (ERD contralateral)')
print('#'*62)
df_mi_r = df[df['condicion'] == 'MI_R']
stat, p = stats.mannwhitneyu(
    df_mi_r['potencia_mu_C3'].values,
    df_mi_r['potencia_mu_C4'].values,
    alternative='less'
)
print(f'\n  Mann-Whitney unilateral (C3 < C4): U={stat:.2f}, p={p:.6f}')
print(f'  Medianas → C3: {df_mi_r["potencia_mu_C3"].median():.4f} | '
      f'C4: {df_mi_r["potencia_mu_C4"].median():.4f}')
print('  → RECHAZA H0 ✓ ERD contralateral confirmado' if p < 0.05
      else '  → NO se rechaza H0')

# ── H2: Beta C3 menor en MI_R que en REST
print('\n' + '#'*62)
print('# HIPÓTESIS 2: Beta C3 — REST vs MI_R')
print('#'*62)
prueba_hipotesis(df, 'REST', 'MI_R', 'potencia_beta_C3')

# ── H3: Mu C4 menor en MM_L que en MI_L
print('\n' + '#'*62)
print('# HIPÓTESIS 3: Mu C4 — MM_L vs MI_L')
print('#'*62)
prueba_hipotesis(df, 'MM_L', 'MI_L', 'potencia_mu_C4')
````

#### *5.4 DIFERENCIAS ENTRE CONDICIONES*

##### **Banda mu (8–13 Hz) — Canal C3**
   
C3 registra la actividad de la corteza motora izquierda, en esta se controla contralateralmente la mano derecha. En reposo se observa la mayor potencia mediana (5.54 µV²/Hz), consistente con una sincronización cortical en ausencia de demanda motora. Al realizar o imaginar movimiento de la mano derecha (MM_R, MI_R), la potencia mu en C3 disminuye notablemente:

- REST → MM_R: reducción del **31 %** en la mediana (5.54 → 3.82 µV²/Hz)

Esta reducción corresponde al lo que se conoce como desincronización relacionada con eventos (ERD), que refleja la activación de la corteza motora contralateral. Es el mecanismo neurofisiológico esperado: cuando la corteza motora izquierda se activa para procesar el movimiento de la mano derecha, las neuronas dejan de oscilar sincrónicamente en mu, reduciendo la potencia de esa banda.

También se observa que el ERD es mayor en movimiento real (MM_R) que en imaginación (MI_R), lo que tiene sentido fisiológico: la ejecución física involucra circuitos adicionales de retroalimentación propioceptiva y salida motora eferente que amplifican la desincronización.

#### *5.6 RESULTADOS DE HIPOTESIS PLANTEADAS INICIALMENTE SOBRE LATERALIZACIÓN Y MODULACIÓN DE LA ACTIVIDAD CORTICAL MOTORA*

##### **Hipótesis 1: Lateralización mu durante Imaginación de Movimiento de Mano Derecha (MI_R)**


**Fundamento neurofisiológico:** La imaginación del movimiento de la mano derecha debe activar preferentemente la corteza motora izquierda (registrada en C3). Si existe ERD contralateral, la potencia mu en C3 debería ser significativamente menor que en C4 durante MI_R.

**Planteamiento de hipótesis:**

Hipotesis Alternativa H1: Durante la imaginación de movimiento de la mano derecha (MI_R), la potencia del ritmo mu (8–13 Hz) en el canal C3 será significativamente menor que en el canal C4.

Hipotesis Nula H0: No existen diferencias significativas en la potencia del ritmo mu entre los canales C3 y C4 durante la condición MI_R.

**Prueba:** Mann-Whitney U unilateral (parte izquierda), ya que la hipótesis alternativa tiene dirección específica.


**Resultados obtenidos:**

| Estadístico | Valor |
|---|---|
| Mediana mu-C3 durante MI_R | 4.3209 µV²/Hz |
| Mediana mu-C4 durante MI_R | 4.9797 µV²/Hz |
| Estadístico U | 2 759 446.00 |
| p-valor | 0.000008 |
| Decisión (α = 0.05) | **Se rechaza H₀** |

    
**Interpretación:** Con p = 8×10⁻⁶ << 0.05, existe evidencia estadística para rechazar H₀. La potencia mu en C3 es significativamente menor que en C4 durante la imaginación de movimiento de mano derecha. Esto confirma la presencia del ERD contralateral en la banda mu, que muestra la activación de la corteza motora izquierda durante la imaginería motora de la mano derecha. Este resultado valida el fundamento fisiológico sobre el cual se construyen los sistemas BCI basados en imaginería motora.

##### **Hipótesis 2 : Diferencia entre reposo e imaginación en mano derecha - canal contralateral (REST vs MI_R)**

**Fundamento neurofisiológico:** Si la imaginación de movimiento de mano derecha activa la corteza motora izquierda, debe producir una reducción de la potencia beta en C3 respecto al estado de reposo (ERD beta).

**Planteamiento de hipótesis:**

Hipotesis Alternativa H1: La potencia del ritmo beta (13-30 Hz) en el canal C3 será significativamente menor durante la imaginación de movimiento de la mano derecha (MI_R) en comparación con la condición de reposo.

Hipotesis Nula H0: No existen diferencias significativas en la potencia del ritmo beta en el canal C3 entre la condición de reposo y la imaginación de movimiento de la mano derecha.

**Prueba:** Mann-Whitney U bilateral (no se asume normalidad; distribuciones confirmadas como no normales por Shapiro-Wilk, p < 0.0001 en ambos grupos).

**Resultados obtenidos:**

| Estadístico | Valor |
|---|---|
| n REST | 9 845 épocas |
| n MI_R | 2 438 épocas |
| Mediana beta-C3 REST | 2.3646 µV²/Hz |
| Mediana beta-C3 MI_R | 1.9213 µV²/Hz |
| Shapiro p (REST / MI_R) | 0.0000 / 0.0000 → NO normal |
| Estadístico U | 13 361 879.00 |
| p-valor | ≈ 0.000000 |
| Decisión (α = 0.05) | **Se rechaza H₀** |

**Interpretación:** El p-valor es prácticamente nulo (p << 0.0001), lo que proporciona evidencia estadística abrumadora para rechazar la hipótesis nula. La potencia beta en C3 durante MI_R (mediana = 1.92 µV²/Hz) es significativamente menor que en REST (mediana = 2.36 µV²/Hz), con una reducción relativa del **19 %** en la mediana. Este resultado confirma que la imaginación de movimiento de mano derecha induce desincronización beta en la corteza motora contralateral (C3), un marcador neurológico bien establecido en la literatura de BCI. La diferencia, aunque estadísticamente robusta, tiene una magnitud moderada en términos absolutos, lo que es esperable dado que la imaginación produce activación cortical de menor intensidad que el movimiento real.

##### **Hipótesis 3 :Diferencia entre el movimiento real de la mano izquierda vs imaginación - canal contralateral : banda mu de C4 (MM_L vs MI_L)**

<div style="text-align: justify;">

**Fundamento neurofisiológico:** Tanto el movimiento real como la imaginación de la mano izquierda activan la corteza motora derecha (C4). ### **6. CONCLUSIONES**

- En conjunto, los datos estadísticos descriptivos evidencian patrones consistentes con la literatura sobre actividad motora en EEG, particularmente la desincronización en las bandas mu y beta durante tareas de movimiento e imaginación.

- Las tres hipótesis planteadas fueron rechazadas con alta significancia estadística (p < 0.0001 en todos los casos), lo que proporciona evidencia sólida de que las diferencias observadas en la potencia espectral entre condiciones no son atribuibles al azar.

- La elección de la prueba no paramétrica Mann-Whitney U fue correcta y necesaria, dado que el test de Shapiro-Wilk rechazó la normalidad en todos los grupos evaluados. El uso de un t-test habría sido inapropiado y podría haber producido inferencias erróneas dada la marcada asimetría de las distribuciones.
  
- En la banda mu, se observa una disminución de la potencia en las condiciones activas (movimiento real e imaginación) en comparación con el estado de reposo, particularmente en el canal C3.Se observa como la reducción de la potencia en esta banda durante la activación motora.

- En el canal C3, asociado al control de la mano derecha, la condición de reposo presenta una media considerablemente mayor que las condiciones de movimiento e imaginación, lo que sugiere una activación cortical durante las tareas. Asimismo, las condiciones de movimiento real (MM) muestran valores ligeramente menores que las de imaginación (MI).
  
- En el canal C4 se da un comportamiento similar al C3, aunque con una dispersión significativamente mayor, que se ve en valores elevados de desviación estándar.
  
- La H1 confirmó el ERD contralateral en mu durante MI_R, la corteza motora izquierda (C3) se desincroniza más que la derecha (C4) al imaginar el movimiento de la mano derecha. 

- La H2 confirmó que la banda beta en C3 disminuye significativamente durante MI_R respecto a REST. Esto demuestra que la desincronización beta es un marcador accesible de la activación cortical motora incluso durante la imaginería, sin necesidad de ejecución física del movimiento.

- La H3 demostró que el movimiento real (MM_L) produce mayor ERD en C4 que la imaginación (MI_L), diferencia que es estadísticamente significativa aunque cuantitativamente moderada (~11 %). Esto tiene implicaciones prácticas importantes.
